```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1) 데이터 로드
df = pd.read_csv("/content/combined_dataset.csv")  # 경로 맞춰서 수정

# 2) 신뢰도 Feature 선택
candidate_features = [
    "review_scores_rating",
    "review_scores_cleanliness",
    "review_scores_communication",
    "review_scores_accuracy",
    "host_response_rate",
    "host_acceptance_rate",
    "number_of_reviews",
    "reviews_per_month",
]
features = [c for c in candidate_features if c in df.columns]

# 3) 퍼센트 문자열 → 숫자
def to_percent_float(s):
    if pd.isna(s): return np.nan
    if isinstance(s, str): return pd.to_numeric(s.replace('%',''), errors='coerce')
    return s

for col in ["host_response_rate", "host_acceptance_rate"]:
    if col in df.columns:
        df[col] = df[col].apply(to_percent_float)

# 4) 결측치 처리: 평점/퍼센트 → 평균, 활동성 변수 → 0
for col in ["review_scores_rating","review_scores_cleanliness","review_scores_communication",
            "review_scores_accuracy","host_response_rate","host_acceptance_rate"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)
        df[col] = df[col].fillna(df[col].mean())

for col in ["number_of_reviews", "reviews_per_month"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0.0)

# 5) 표준화
X = df[features].copy()
scaler = StandardScaler()
Xz = scaler.fit_transform(X)
Xz_df = pd.DataFrame(Xz, columns=features, index=df.index)

# 신뢰도 종합 z-점수(피처 평균)
df["trust_composite_z"] = Xz_df.mean(axis=1)

# 6) KMeans (k=4)
kmeans = KMeans(n_clusters=4, n_init=20, random_state=42)
df["trust_cluster_id"] = kmeans.fit_predict(Xz_df)

# 7) 등급 매핑: 클러스터별 종합점 평균으로 순위→등급
cluster_rank = (
    df.groupby("trust_cluster_id")["trust_composite_z"]
      .mean().sort_values()              # 낮은 → 높은
)
ordered_ids = list(cluster_rank.index)   # 예: [2, 1, 3, 0]

levels = ["루키", "라이징", "슈퍼", "마스터"]
id2level = {cid: levels[i] for i, cid in enumerate(ordered_ids)}
df["trust_level"] = df["trust_cluster_id"].map(id2level)

# 8) 등급별 요약
summary_cols = [c for c in [
    "review_scores_rating","review_scores_cleanliness","review_scores_communication",
    "review_scores_accuracy","host_response_rate","host_acceptance_rate",
    "number_of_reviews","reviews_per_month","trust_composite_z"
] if c in df.columns]

summary = (
    df.groupby("trust_level")[summary_cols]
      .mean().reindex(levels).round(3)
)
counts = df["trust_level"].value_counts().reindex(levels).fillna(0).astype(int)
summary.insert(0, "개수", counts.values)
print(summary)

# 9) 결과 저장
cols_keep = ["trust_level","trust_cluster_id","trust_composite_z"] + features
if "id" in df.columns: cols_keep = ["id"] + cols_keep
if "host_id" in df.columns: cols_keep = ["host_id"] + cols_keep

df[cols_keep].to_csv("host_trust_levels_labeled.csv", index=False)
print("Saved:", "host_trust_levels_labeled.csv")
```

```python
# ===========================
# 1) 등급별 요약/사분위 표
# ===========================
import pandas as pd
import numpy as np

summary_cols = [c for c in [
    "review_scores_rating","review_scores_cleanliness","review_scores_communication",
    "review_scores_accuracy","host_response_rate","host_acceptance_rate",
    "number_of_reviews","reviews_per_month","trust_composite_z"
] if c in df.columns]

# 평균 요약
level_order = ["루키","라이징","슈퍼","마스터"]
summary = (
    df.groupby("trust_level")[summary_cols]
      .mean().reindex(level_order).round(3)
)
counts = df["trust_level"].value_counts().reindex(level_order).fillna(0).astype(int)
summary.insert(0, "개수", counts.values)
print("\n[등급별 평균 프로파일]")
print(summary)

# 사분위(0.25/0.5/0.75) → 각 등급의 대략적 기준 범위 확인
quantiles = (
    df.groupby("trust_level")[ [c for c in summary_cols if c != "trust_composite_z"] ]
      .quantile([0.25, 0.5, 0.75])
      .unstack(level=-1)
      .reindex(level_order)
      .round(2)
)
print("\n[등급별 사분위(0.25/0.5/0.75) 표]")
print(quantiles)

# ===========================
# 2) 레이더 차트 (표준화 평균)
# ===========================
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from math import pi

radar_feats = [c for c in [
    "review_scores_rating","review_scores_cleanliness","review_scores_communication",
    "review_scores_accuracy","host_response_rate","host_acceptance_rate",
    "number_of_reviews","reviews_per_month"
] if c in df.columns]

scaler_temp = StandardScaler()
Xz = scaler_temp.fit_transform(df[radar_feats])
Xz_df = pd.DataFrame(Xz, columns=radar_feats, index=df.index)

radar_means = (
    Xz_df.assign(level=df["trust_level"])
         .groupby("level")[radar_feats].mean()
         .reindex(level_order)
)

N = len(radar_feats)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)
for lvl in level_order:
    vals = radar_means.loc[lvl].tolist()
    vals += vals[:1]
    ax.plot(angles, vals, linewidth=2, label=lvl)
    ax.fill(angles, vals, alpha=0.1)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(radar_feats, fontsize=9)
ax.set_yticklabels([])
ax.set_title("신뢰도 등급별 표준화 평균(레이더)")
plt.legend()
plt.tight_layout()
plt.show()

# ===========================
# 3) 간단 산점도 2종
# ===========================
def scatter_by_level(df, x, y, levels):
    plt.figure(figsize=(7,5))
    for lvl in levels:
        sub = df[df["trust_level"] == lvl]
        plt.scatter(sub[x], sub[y], s=12, alpha=0.7, label=lvl)
    plt.xlabel(x); plt.ylabel(y)
    plt.title(f"등급별 분포: {x} vs {y}")
    plt.legend()
    plt.tight_layout()
    plt.show()

if "reviews_per_month" in df.columns and "review_scores_rating" in df.columns:
    scatter_by_level(df, "reviews_per_month", "review_scores_rating", level_order)

if "number_of_reviews" in df.columns and "review_scores_rating" in df.columns:
    scatter_by_level(df, "number_of_reviews", "review_scores_rating", level_order)

# ===========================
# 4) PCA 2D 시각화 (전체 피처 기반)
# ===========================
from sklearn.decomposition import PCA

pca_feats = [c for c in radar_feats if c in df.columns]
pca = PCA(n_components=2, random_state=42)
Z = pca.fit_transform(df[pca_feats])

plt.figure(figsize=(7,5))
for lvl in level_order:
    idx = (df["trust_level"] == lvl).values
    plt.scatter(Z[idx,0], Z[idx,1], s=12, alpha=0.7, label=lvl)
plt.xlabel("PCA1"); plt.ylabel("PCA2")
plt.title("신뢰도 피처 기반 PCA 2D 시각화")
plt.legend()
plt.tight_layout()
plt.show()

# ===========================
# 5) 박스플롯 (trust_composite_z)
# ===========================
plt.figure(figsize=(7,5))
df.boxplot(column="trust_composite_z", by="trust_level")
plt.suptitle("")
plt.title("등급별 신뢰도 종합 z-점수 분포")
plt.xlabel("신뢰도 등급"); plt.ylabel("trust_composite_z")
plt.tight_layout()
plt.show()

# ===========================
# 6) '사람이 읽을 수 있는' 분할 기준 추출
#    (의사결정나무로 등급 라벨을 설명)
# ===========================
from sklearn.tree import DecisionTreeClassifier, export_text

rule_feats = [c for c in [
    "review_scores_rating","review_scores_cleanliness","review_scores_communication",
    "review_scores_accuracy","host_response_rate","host_acceptance_rate",
    "number_of_reviews","reviews_per_month"
] if c in df.columns]

X_rule = df[rule_feats].copy()
y_rule = df["trust_level"].copy()

dt = DecisionTreeClassifier(max_depth=3, min_samples_leaf=50, random_state=42)
dt.fit(X_rule, y_rule)

rules = export_text(dt, feature_names=rule_feats, decimals=2)
print("\n=== 해석용 분할 규칙(의사결정나무, 최대깊이=3) ===")
print(rules)

# 변수 중요도(참고)
imp = pd.Series(dt.feature_importances_, index=rule_feats).sort_values(ascending=False)
print("\n[의사결정나무 기준 변수 중요도]")
print(imp.round(3))
```

```python
import pandas as pd
import numpy as np

# ==== 0) 데이터 로드 ====
df = pd.read_csv("combined_dataset.csv")  # 경로 수정

# ==== 1) 필요한 컬럼만 사용 & 전처리 ====
FEATURES = ["number_of_reviews", "host_response_rate", "host_listings_count", "review_scores_rating"]
missing = [c for c in FEATURES if c not in df.columns]
if missing:
    raise ValueError(f"데이터에 다음 컬럼이 없습니다: {missing}")

# host_response_rate가 "95%" 같은 문자열일 수 있으니 숫자로 변환
def pct_to_float(x):
    if pd.isna(x): return np.nan
    if isinstance(x, str): return pd.to_numeric(x.replace("%",""), errors="coerce")
    return x

df["host_response_rate"] = df["host_response_rate"].apply(pct_to_float)

# 숫자형 변환
for c in FEATURES:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# 결측치 처리: 응답률/평점은 평균으로, 카운트형은 0으로
df["host_response_rate"]  = df["host_response_rate"].fillna(df["host_response_rate"].mean())
df["review_scores_rating"] = df["review_scores_rating"].fillna(df["review_scores_rating"].mean())
df["number_of_reviews"]    = df["number_of_reviews"].fillna(0)
df["host_listings_count"]  = df["host_listings_count"].fillna(0)

# ==== 2) 지표별 기준점(사분위수) 계산 ====
Q = df[FEATURES].quantile([0.25, 0.5, 0.75]).round(3)  # Q1, Q2, Q3
print("\n[지표별 등급 기준점(사분위수)]")
print(Q.rename(index={0.25:"Q1(루키/라이징 경계)", 0.5:"Q2(라이징/슈퍼 경계)", 0.75:"Q3(슈퍼/마스터 경계)"}))

# ==== 3) 단일 지표 → 4등급 매핑 함수 (사분위수 기반) ====
# 값 < Q1 → 루키(0), Q1~Q2 → 라이징(1), Q2~Q3 → 슈퍼(2), ≥Q3 → 마스터(3)
def to_level_by_quantile(val, q1, q2, q3):
    if pd.isna(val): return 0
    if val < q1:   return 0
    if val < q2:   return 1
    if val < q3:   return 2
    return 3

# ==== 4) 행별 등급 산정 로직 ====
# 각 지표에서 등급(0~3)을 만들고, 평균 등급을 반올림해서 최종 등급으로 사용
# (지표 가중치를 다르게 주고 싶다면 weights 딕셔너리로 수정)
weights = {
    "number_of_reviews":   1.0,
    "host_response_rate":  1.0,
    "host_listings_count": 1.0,
    "review_scores_rating":1.0,
}
total_w = sum(weights.values())

q1 = Q.loc[0.25]; q2 = Q.loc[0.5]; q3 = Q.loc[0.75]

def assign_level(row):
    lv = {}
    for c in FEATURES:
        lv[c] = to_level_by_quantile(row[c], q1[c], q2[c], q3[c])
    # 가중 평균 → 최종 등급(0~3)으로 반올림
    avg_level = sum(lv[c]*weights[c] for c in FEATURES) / total_w
    final_lvl = int(round(avg_level))
    return final_lvl, lv  # 최종등급, 지표별 등급도 같이 반환

levels_map = {0:"루키", 1:"라이징", 2:"슈퍼", 3:"마스터"}

final_levels = []
per_feature_levels = []
for _, r in df.iterrows():
    fl, lv_dict = assign_level(r)
    final_levels.append(fl)
    per_feature_levels.append(lv_dict)

df["trust_level_id"] = final_levels
df["trust_level"] = df["trust_level_id"].map(levels_map)

# 지표별 등급(0~3)을 컬럼으로 저장(선택)
for c in FEATURES:
    df[f"level_{c}"] = [d[c] for d in per_feature_levels]

# ==== 5) 결과 요약 ====
print("\n[최종 등급 분포]")
print(df["trust_level"].value_counts().reindex(["루키","라이징","슈퍼","마스터"]).fillna(0).astype(int))

print("\n[등급별 평균(핵심 지표)]")
print(
    df.groupby("trust_level")[FEATURES]
      .mean().reindex(["루키","라이징","슈퍼","마스터"]).round(2)
)

# ==== 6) 실제 ‘기준점’을 사람이 읽도록 예시 출력 ====
def pretty_rule(name, unit, q1v, q2v, q3v):
    return (f"- {name}: "
            f"루키 < {q1v}{unit} ≤ 라이징 < {q2v}{unit} ≤ 슈퍼 < {q3v}{unit} ≤ 마스터")

print("\n[사람이 읽을 수 있는 기준 요약]")
print(pretty_rule("리뷰 수(number_of_reviews)", "개", int(q1["number_of_reviews"]), int(q2["number_of_reviews"]), int(q3["number_of_reviews"])))
print(pretty_rule("응답률(host_response_rate)", "%", round(q1["host_response_rate"],1), round(q2["host_response_rate"],1), round(q3["host_response_rate"],1)))
print(pretty_rule("호스팅 횟수(host_listings_count)", "개", int(q1["host_listings_count"]), int(q2["host_listings_count"]), int(q3["host_listings_count"])))
print(pretty_rule("평점(review_scores_rating)", "점", round(q1["review_scores_rating"],2), round(q2["review_scores_rating"],2), round(q3["review_scores_rating"],2)))

# ==== 7) 저장(선택) ====
keep_cols = ["trust_level","trust_level_id"] + FEATURES + [f"level_{c}" for c in FEATURES]
if "id" in df.columns: keep_cols = ["id"] + keep_cols
if "host_id" in df.columns: keep_cols = ["host_id"] + keep_cols
df[keep_cols].to_csv("trust_levels_by_rules.csv", index=False)
print("\nSaved: trust_levels_by_rules.csv")
```

```python
# === 기본 설치 (필요 시) ===
# pip install pandas numpy scikit-learn

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, export_text

# -----------------------------
# 0) 데이터 로드 & 컬럼 지정
# -----------------------------
PATH = "/content/combined_dataset.csv"  # 경로 맞춰 수정
df = pd.read_csv(PATH)

# 사용 지표(딱 4개)
FEATURES = ["number_of_reviews", "host_response_rate", "host_listings_count", "review_scores_rating"]
missing = [c for c in FEATURES if c not in df.columns]
if missing:
    raise ValueError(f"다음 컬럼이 없습니다: {missing}")

# 퍼센트 문자열 → 숫자 (예: "95%")
def pct_to_float(x):
    if pd.isna(x): return np.nan
    if isinstance(x, str): return pd.to_numeric(x.replace("%",""), errors="coerce")
    return x

df["host_response_rate"] = df["host_response_rate"].apply(pct_to_float)

# 숫자 변환 & 결측치 처리 (간단 규칙)
for c in FEATURES:
    df[c] = pd.to_numeric(df[c], errors="coerce")
# 응답률/평점 → 평균 대체, 카운트형 → 0
df["host_response_rate"]  = df["host_response_rate"].fillna(df["host_response_rate"].mean())
df["review_scores_rating"] = df["review_scores_rating"].fillna(df["review_scores_rating"].mean())
df["number_of_reviews"]    = df["number_of_reviews"].fillna(0)
df["host_listings_count"]  = df["host_listings_count"].fillna(0)

# -----------------------------
# 1) 표준화 + KMeans(k=4)
# -----------------------------
scaler = StandardScaler()
X = df[FEATURES].copy()
Xz = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=4, n_init=20, random_state=42)
df["cluster_id"] = kmeans.fit_predict(Xz)

# -----------------------------
# 2) 등급명 부여 (루키→라이징→슈퍼→마스터)
#    - 클러스터별 '신뢰도 종합 z-점수' 평균으로 서열화
# -----------------------------
trust_z = Xz.mean(axis=1)
df["trust_z_mean"] = trust_z
order = (
    df.groupby("cluster_id")["trust_z_mean"]
      .mean().sort_values().index.tolist()  # 낮은 → 높은
)
LEVELS = ["루키","라이징","슈퍼","마스터"]
id2level = {cid: LEVELS[i] for i, cid in enumerate(order)}
df["trust_level"] = df["cluster_id"].map(id2level)

print("\n[등급 분포]")
print(df["trust_level"].value_counts().reindex(LEVELS).fillna(0).astype(int))

# -----------------------------
# 3) '원래 단위'에서의 클러스터 기준 요약
#    - (a) 클러스터 중심(centroid) 역변환 → 대표값
#    - (b) 클러스터별 사분위(0.25/0.5/0.75) → 범위
# -----------------------------
# (a) 중심값(원 단위)
centroids_z = kmeans.cluster_centers_                     # 표준화 공간
centroids_orig = pd.DataFrame(scaler.inverse_transform(centroids_z),
                              columns=FEATURES)
centroids_orig["trust_level"] = [id2level[cid] for cid in range(4)]
centroids_orig = centroids_orig.set_index("trust_level").loc[LEVELS].round(2)

print("\n[등급별 대표값(클러스터 중심, 원래 단위)]")
print(centroids_orig)

# (b) 사분위로 범위 제시
qtab = (
    df.groupby("trust_level")[FEATURES]
      .quantile([0.25, 0.5, 0.75])
      .unstack(level=-1)
      .reindex(LEVELS)
      .round(2)
)
qtab.columns = [f"{col}_Q{int(q*100)}" for col,q in qtab.columns]  # 보기 좋게
print("\n[등급별 범위(사분위, 원래 단위)]")
print(qtab)

# -----------------------------
# 4) 사람이 읽을 수 있는 '분할 기준' 자동 추출
#    - 의사결정나무를 원 단위 피처로 학습해 간단 규칙 출력
# -----------------------------
X_rule = df[FEATURES].copy()
y_rule = df["trust_level"].copy()
tree = DecisionTreeClassifier(max_depth=3, min_samples_leaf=50, random_state=42)
tree.fit(X_rule, y_rule)
rules = export_text(tree, feature_names=FEATURES, decimals=2)
print("\n=== 해석용 분할 규칙(의사결정나무, 깊이<=3) ===")
print(rules)

# 중요 변수 참고
importances = pd.Series(tree.feature_importances_, index=FEATURES).sort_values(ascending=False)
print("\n[분할에 기여한 변수 중요도]")
print(importances.round(3))

# -----------------------------
# 5) 결과 저장(선택)
# -----------------------------
out_csv = "trust_levels_kmeans4.csv"
keep = ["trust_level", "cluster_id", "trust_z_mean"] + FEATURES
if "id" in df.columns: keep = ["id"] + keep
if "host_id" in df.columns: keep = ["host_id"] + keep
df[keep].to_csv(out_csv, index=False)
print(f"\nSaved: {out_csv}")
```

```python
import pandas as pd
import numpy as np

# 전제: df에 이미 trust_level(루키/라이징/슈퍼/마스터) 라벨이 붙어 있음
# (앞서 KMeans 후 trust_level을 매핑한 상태에서 실행)

FEATURES = ["number_of_reviews", "host_response_rate", "host_listings_count", "review_scores_rating"]
LEVELS  = ["루키","라이징","슈퍼","마스터"]

# 1) 등급별(원 단위) 평균 = 군집 대표값
centroids = (
    df.groupby("trust_level")[FEATURES]
      .mean().reindex(LEVELS)
)

# 2) 인접 등급 간 "중간값" = 전환 기준(Threshold)
#    예: 루키(평균 3개) ↔ 라이징(평균 10개) → 리뷰수 기준은 (3+10)/2 = 6.5
def midpoint(a, b):
    return (a + b) / 2.0

thresholds = {}
for i in range(len(LEVELS)-1):
    low = LEVELS[i]
    high = LEVELS[i+1]
    row = {}
    for feat in FEATURES:
        row[feat] = midpoint(centroids.loc[low, feat], centroids.loc[high, feat])
    thresholds[f"{low}→{high}"] = row

thresh_df = pd.DataFrame(thresholds).T.round(3)
print("\n[등급 전환 기준(인접 등급 평균의 중간점, 원 단위)]")
print(thresh_df)

# 3) 사람이 읽기 좋은 문장으로 출력 (증가/감소 방향 자동 표기)
def sign_symbol(v_from, v_to):
    # 상위 등급에서 값이 더 크면 '≥', 더 작으면 '≤'로 안내
    return "≥" if (v_to - v_from) >= 0 else "≤"

print("\n[등급 전환 한 줄 요약]")
for i in range(len(LEVELS)-1):
    low = LEVELS[i]
    high = LEVELS[i+1]
    parts = []
    for feat in FEATURES:
        thr   = thresholds[f"{low}→{high}"][feat]
        sym   = sign_symbol(centroids.loc[low, feat], centroids.loc[high, feat])
        unit  = "개" if feat in ["number_of_reviews","host_listings_count"] else "%"
        if feat == "review_scores_rating":
            unit = "점"
        label = {
            "number_of_reviews":"리뷰수",
            "host_response_rate":"응답률",
            "host_listings_count":"호스팅 횟수",
            "review_scores_rating":"평점"
        }[feat]
        parts.append(f"{label}: {sym} {thr:.2f}{unit}")
    print(f"{low} → {high} 기준: " + ", ".join(parts))

# 4) (선택) CSV로 저장
thresh_df.to_csv("trust_transition_thresholds.csv")
print("\nSaved: trust_transition_thresholds.csv")
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np

# ===== 경로 설정 =====
INPUT_PATH  = "/content/combined_dataset.csv"            # 파일 경로에 맞게 수정
OUTPUT_HOST = "/content/host_aggregated_quartiles.csv"   # 호스트 단위 결과
# =====================

# 1) 데이터 로드
df = pd.read_csv(INPUT_PATH)

# 2) 실제 사용하는 컬럼들 (데이터셋 기준)
REQ_COLS = [
    "host_id",
    "number_of_reviews",
    "review_scores_rating",
    "host_response_rate",
    "host_acceptance_rate",
]
missing = [c for c in REQ_COLS if c not in df.columns]
if missing:
    raise KeyError(f"데이터에 없는 컬럼: {missing}")

# 3) 수치 변환
def to_num(series: pd.Series) -> pd.Series:
    """ %, , 제거 후 숫자로 변환 """
    return pd.to_numeric(
        series.astype(str)
              .str.replace("%", "", regex=False)
              .str.replace(",", "", regex=False)
              .str.strip(),
        errors="coerce"
    )

df["number_of_reviews"]    = to_num(df["number_of_reviews"])
df["review_scores_rating"] = to_num(df["review_scores_rating"])
df["host_response_rate"]   = to_num(df["host_response_rate"])
df["host_acceptance_rate"] = to_num(df["host_acceptance_rate"])

# host_response_rate / host_acceptance_rate가 0~1 범위라면 0~100으로 보정
for rate_col in ["host_response_rate", "host_acceptance_rate"]:
    ser = df[rate_col].dropna()
    if len(ser) > 0 and ser.between(0, 1).all():
        df[rate_col] = df[rate_col] * 100.0

# 4) 호스트 단위 집계
#    - 리뷰수: 합(sum) (호스트가 가진 모든 리스팅의 리뷰수 총합)
#    - 평균평점: 평균(mean)
#    - 응답률/수락률: 평균(mean)
agg_df = (
    df.groupby("host_id", as_index=False)
      .agg(
          total_reviews=("number_of_reviews", "sum"),
          avg_rating=("review_scores_rating", "mean"),
          avg_response_rate=("host_response_rate", "mean"),
          avg_acceptance_rate=("host_acceptance_rate", "mean"),
      )
)

# 5) 사분위 라벨 함수 (Q1~Q4 → 1~4)
def quartile_labels(s: pd.Series) -> pd.Series:
    s = s.copy()
    # 고유값 적을 때를 대비해 안전 처리
    try:
        q = pd.qcut(s.rank(method="first"), 4, labels=[1, 2, 3, 4])
        return q.astype("Int64")
    except Exception:
        s_clean = s.dropna()
        if s_clean.nunique() < 2:
            return pd.Series([pd.NA] * len(s), index=s.index, dtype="Int64")
        bins = np.unique(np.nanpercentile(s_clean, [0, 25, 50, 75, 100]))
        if len(bins) < 3:
            return pd.Series([pd.NA] * len(s), index=s.index, dtype="Int64")
        cat = pd.cut(s, bins=bins, include_lowest=True, labels=False)
        if cat is None:
            return pd.Series([pd.NA] * len(s), index=s.index, dtype="Int64")
        return (cat + 1).astype("Int64")

# 6) 지표별 사분위 라벨 추가 (1=낮음, 4=높음)
agg_df["리뷰수_사분위(1-낮음,4-높음)"]   = quartile_labels(agg_df["total_reviews"])
agg_df["평균평점_사분위(1-낮음,4-높음)"] = quartile_labels(agg_df["avg_rating"])
agg_df["응답률_사분위(1-낮음,4-높음)"]   = quartile_labels(agg_df["avg_response_rate"])
agg_df["수락률_사분위(1-낮음,4-높음)"]   = quartile_labels(agg_df["avg_acceptance_rate"])

# 7) 지표별 분포 요약(선택)
def summary_by_quartile(col_name: str) -> pd.DataFrame:
    counts = agg_df[col_name].value_counts(dropna=False).sort_index()
    total = counts.sum()
    return pd.DataFrame({
        "사분위": ["Q1","Q2","Q3","Q4","NA"][:len(counts)],
        "개수": counts.astype(int).tolist(),
        "비율(%)": [round((c/total)*100, 2) if total else 0 for c in counts]
    })

# 8) 저장
agg_df.to_csv(OUTPUT_HOST, index=False, encoding="utf-8-sig")

print(f"[완료] 호스트 단위 사분위 라벨 결과 저장: {OUTPUT_HOST}")
print(agg_df.head(10))
print("\n[요약] 리뷰수 분포\n", summary_by_quartile("리뷰수_사분위(1-낮음,4-높음)"))
print("\n[요약] 평균평점 분포\n", summary_by_quartile("평균평점_사분위(1-낮음,4-높음)"))
print("\n[요약] 응답률 분포\n", summary_by_quartile("응답률_사분위(1-낮음,4-높음)"))
print("\n[요약] 수락률 분포\n", summary_by_quartile("수락률_사분위(1-낮음,4-높음)"))
```

```python
# 6-1) 사분위 경계값 확인 함수
def quartile_summary(s: pd.Series, name: str):
    q = s.quantile([0, 0.25, 0.5, 0.75, 1.0])
    print(f"\n[{name}] 사분위 기준값")
    print(f" - 최소값: {q.loc[0]:.2f}")
    print(f" - Q1(25%): {q.loc[0.25]:.2f}")
    print(f" - 중앙값(Q2): {q.loc[0.5]:.2f}")
    print(f" - Q3(75%): {q.loc[0.75]:.2f}")
    print(f" - 최대값: {q.loc[1.0]:.2f}")

# 6-2) 각 지표별 경계 출력
quartile_summary(agg_df["total_reviews"], "리뷰수")
quartile_summary(agg_df["avg_rating"], "평균평점")
quartile_summary(agg_df["avg_response_rate"], "응답률")
quartile_summary(agg_df["avg_acceptance_rate"], "예약 수락률")
```

```python
import pandas as pd, numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, export_text

# 1) 타깃(booking_rate) 생성
df["booking_rate"] = (df["reviews_per_month"]*30 / df["availability_365"]).clip(0,1)
df = df.replace([np.inf,-np.inf], np.nan).dropna(subset=["booking_rate"])

# 2) 후보 지표
feats = ["number_of_reviews","host_response_rate","host_listings_count","review_scores_rating"]
# 전처리: % -> 숫자, 결측/이상치 처리
df["host_response_rate"] = (
    df["host_response_rate"].astype(str).str.replace("%","",regex=False).astype(float)
)
for c in feats:
    df[c] = pd.to_numeric(df[c], errors="coerce")
df = df.dropna(subset=feats)

# 3) 중요도(근거) 추정
X = df[feats].copy()
y = df["booking_rate"].values
rf = RandomForestRegressor(n_estimators=500, random_state=42)
rf.fit(X, y)
perm = permutation_importance(rf, X, y, n_repeats=10, random_state=42)
imp = pd.Series(perm.importances_mean, index=feats).clip(lower=0)  # 음수면 0으로
weights = (imp / imp.sum()).fillna(0)
print("[중요도 기반 가중치]\n", weights.sort_values(ascending=False).round(3))

# 4) 방향 정렬(타깃과 상관 부호)
signs = np.sign(pd.Series([np.corrcoef(df[c], y)[0,1] for c in feats], index=feats)).replace(0,1)

# 5) 가중 점수 → 표준화 + 합산
scaler = StandardScaler()
Z = pd.DataFrame(scaler.fit_transform(X), columns=feats, index=df.index)
Z_signed = Z.mul(signs, axis=1)
df["score"] = (Z_signed * weights).sum(axis=1)

# 6) 4등급 분할(사분위)
q = df["score"].quantile([0.25,0.5,0.75])
def to_level(s):
    if s<q.iloc[0]: return "루키"
    if s<q.iloc[1]: return "라이징"
    if s<q.iloc[2]: return "슈퍼"
    return "마스터"
df["level_quantile"] = df["score"].apply(to_level)

# 7) KMeans로 교차 체크
km = KMeans(n_clusters=4, n_init=20, random_state=42).fit(Z_signed)
df["cluster_id"] = km.labels_
order = df.groupby("cluster_id")["score"].mean().sort_values().index
map2lvl = {cid: ["루키","라이징","슈퍼","마스터"][i] for i,cid in enumerate(order)}
df["level_kmeans"] = df["cluster_id"].map(map2lvl)

print("\n[사분위 vs KMeans 교차표]\n", pd.crosstab(df["level_quantile"], df["level_kmeans"]))

# 8) 규칙화(운영 컷): 트리로 원 단위 규칙 도출
dt = DecisionTreeClassifier(max_depth=3, min_samples_leaf=50, random_state=42)
dt.fit(X, df["level_quantile"])
rules = export_text(dt, feature_names=feats, decimals=2)
print("\n[등급 규칙(원 단위 컷오프)]\n", rules)

# 9) 단조성 점검(등급↑ → booking_rate↑?)
print("\n[등급별 예약률 평균]\n",
      df.groupby("level_quantile")["booking_rate"].mean().reindex(["루키","라이징","슈퍼","마스터"]).round(3))
```

```python
import scipy.stats as stats

# 등급별 booking_rate 평균 비교
group_means = df.groupby("level_quantile")["booking_rate"].mean()
anova_p = stats.f_oneway(
    *[df.loc[df["level_quantile"]==lvl, "booking_rate"] for lvl in ["루키","라이징","슈퍼","마스터"]]
).pvalue

print("[등급별 예약률 평균]")
print(group_means)
print(f"ANOVA p-value = {anova_p:.4f}")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# === 0) 데이터 로드 ===
df = pd.read_csv("/content/combined_dataset.csv")

# === 1) 전처리: amenity_count, 타입 표준화 ===
# amenities → 쉼표 기준 개수
df["amenity_count"] = (
    df["amenities"].fillna("").astype(str)
      .apply(lambda x: len([s for s in x.split(",") if s.strip()]))
)

# host_is_superhost → 0/1로 표준화
df["host_is_superhost_norm"] = (
    df["host_is_superhost"]
      .astype(str).str.lower().str.strip()
      .map({"t":1,"true":1,"1":1,"y":1,"yes":1,"f":0,"false":0,"0":0,"n":0,"no":0})
)

# 숫자 컬럼 강제 변환
for col in ["review_scores_rating","number_of_reviews"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# === 2) 슈퍼호스트만 필터 + 유효표본 보정 ===
need_cols = ["review_scores_rating","number_of_reviews","amenity_count","host_is_superhost_norm"]
missing = [c for c in need_cols if c not in df.columns]
if missing:
    raise ValueError(f"필요 컬럼이 없습니다: {missing}")

df_super = df[df["host_is_superhost_norm"] == 1].copy()

# 타깃 결측 제거
df_super = df_super.dropna(subset=["review_scores_rating"])

# 두 설명변수 결측 제거
df_super = df_super.dropna(subset=["number_of_reviews","amenity_count"])

print(f"[INFO] 슈퍼호스트 표본수: {len(df_super)}")
if len(df_super) < 30:
    print("[WARN] 슈퍼호스트 유효 표본이 30건 미만입니다. 통계적 신뢰도가 낮을 수 있습니다.")

# 분산 0(상수열) 체크 → 상관/회귀 불가 방지
def has_variance(s):
    s = pd.to_numeric(s, errors="coerce").dropna()
    return s.nunique() > 1

ok_reviews = has_variance(df_super["number_of_reviews"])
ok_amen = has_variance(df_super["amenity_count"])
ok_y = has_variance(df_super["review_scores_rating"])

if not (ok_reviews and ok_amen and ok_y):
    print("[WARN] 한 변수 이상이 분산이 0이거나 유효 값 부족입니다.")
    print(f" - number_of_reviews variance>0? {ok_reviews}")
    print(f" - amenity_count variance>0?   {ok_amen}")
    print(f" - review_scores_rating var>0? {ok_y}")

# === 3) 상관계수 비교 (슈퍼호스트 한정) ===
corr_reviews = df_super["number_of_reviews"].corr(df_super["review_scores_rating"]) if ok_reviews and ok_y else np.nan
corr_amenities = df_super["amenity_count"].corr(df_super["review_scores_rating"])     if ok_amen and ok_y else np.nan
print(f"슈퍼호스트 내 리뷰수-평점 상관계수: {corr_reviews:.3f}" if pd.notna(corr_reviews) else "리뷰수-평점 상관계수: 계산 불가")
print(f"슈퍼호스트 내 어메니티-평점 상관계수: {corr_amenities:.3f}" if pd.notna(corr_amenities) else "어메니티-평점 상관계수: 계산 불가")

# === 4) 표준화 회귀 (두 변수 동시 투입, 표준화 계수 비교) ===
# 표준화: 분산 0 방지용으로 std=0이면 해당 컬럼 제외
X = df_super[["number_of_reviews","amenity_count"]].copy()
y = df_super["review_scores_rating"].copy()

# 표준화 (안전 처리)
std = X.std(ddof=0).replace(0, np.nan)
Xz = (X - X.mean()) / std
use_cols = [c for c in Xz.columns if Xz[c].notna().all()]
Xz = Xz[use_cols].dropna()
y_fit = y.loc[Xz.index]

# 회귀 적합 (가능한 경우에만)
if len(use_cols) >= 1 and y_fit.notna().sum() >= max(30, len(use_cols)*10):
    Xz_const = sm.add_constant(Xz)
    model = sm.OLS(y_fit, Xz_const).fit()
    print(model.summary())
    # 표준화 계수(절대값) 비교로 영향력 판단
    beta = model.params.drop("const", errors="ignore").abs().sort_values(ascending=False)
    print("\n[표준화 계수(절대값) Top]:")
    print(beta)
else:
    print("[WARN] 표준화 회귀를 돌릴 충분한 유효 표본이 없거나, 설명변수 분산 문제로 제외되었습니다.")
    print(f"사용가능 변수: {use_cols}, 유효 표본수: {y_fit.notna().sum()}")

# === 5) 시각화 (슈퍼호스트 한정 산점도+회귀선) ===
fig, axes = plt.subplots(1, 2, figsize=(12,5))
sns.regplot(data=df_super, x="number_of_reviews", y="review_scores_rating",
            scatter_kws={"alpha":0.3}, ax=axes[0])
axes[0].set_title("슈퍼호스트: 리뷰수 vs 평점")

sns.regplot(data=df_super, x="amenity_count", y="review_scores_rating",
            scatter_kws={"alpha":0.3}, ax=axes[1])
axes[1].set_title("슈퍼호스트: 어메니티 수 vs 평점")

plt.tight_layout()
plt.show()
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# 0) 데이터 로드 -------------------------------------------------------------
df = pd.read_csv("/content/combined_dataset.csv")

# 1) 전처리: amenity_count, host_is_superhost 표준화 -------------------------
# amenities → 쉼표 기준 개수
df["amenity_count"] = (
    df["amenities"].fillna("").astype(str)
      .apply(lambda x: len([s for s in x.split(",") if s.strip()]))
)

# host_is_superhost → 0/1 정규화(다양한 표기 대응)
if "host_is_superhost" in df.columns:
    sup_map = {"t":1,"true":1,"y":1,"yes":1,"1":1,
               "f":0,"false":0,"n":0,"no":0,"0":0}
    df["host_is_superhost"] = (
        df["host_is_superhost"].astype(str).str.lower().str.strip().map(sup_map)
    )

# 숫자형으로 안전 변환(필요 컬럼 위주)
num_like_cols = [
    "price","number_of_reviews","reviews_per_month",
    "amenity_count","availability_365","host_is_superhost"
]
for c in num_like_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# 2) 학습용 X, y 구성 --------------------------------------------------------
target = "review_scores_rating"
if target not in df.columns:
    raise ValueError("review_scores_rating 컬럼이 없습니다.")

mask_notna = df[target].notna()
X_all = df.loc[:, num_like_cols].copy()
y_all = df.loc[:, target].astype(float)

# 3) 결측 처리 전략 정의(칼럼 특성 고려) -------------------------------------
# - price, availability_365: 중앙값 대체 + 결측 인디케이터(추가 피처)
median_with_indicator_cols = [c for c in ["price", "availability_365"] if c in X_all.columns]

# - number_of_reviews, reviews_per_month, amenity_count, host_is_superhost:
#   0이 의미 있는 경우가 많아 0으로 대체
zero_fill_cols = [c for c in ["number_of_reviews","reviews_per_month","amenity_count","host_is_superhost"]
                  if c in X_all.columns]

preprocess = ColumnTransformer(
    transformers=[
        ("med_ind", SimpleImputer(strategy="median", add_indicator=True), median_with_indicator_cols),
        ("zero",    SimpleImputer(strategy="constant", fill_value=0),     zero_fill_cols),
    ],
    remainder="drop",  # 정의되지 않은 컬럼은 사용하지 않음
)

# 4) 모델 파이프라인 ----------------------------------------------------------
model = Pipeline(steps=[
    ("prep", preprocess),
    ("rf", RandomForestRegressor(
        n_estimators=200,
        max_depth=12,
        min_samples_leaf=5,
        n_jobs=-1,
        random_state=42
    ))
])

# 5) 학습/검증 분할 & 학습 ----------------------------------------------------
X_train, X_val, y_train, y_val = train_test_split(
    X_all.loc[mask_notna], y_all.loc[mask_notna],
    test_size=0.2, random_state=42
)

model.fit(X_train, y_train)

# 6) 성능 점검(검증셋) --------------------------------------------------------
y_pred = model.predict(X_val)
r2 = r2_score(y_val, y_pred)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f"[Validation] R² = {r2:.3f} | RMSE = {rmse:.3f}")

# 7) 결측 타깃 예측 & 채우기 --------------------------------------------------
mask_na = df[target].isna()
if mask_na.any():
    y_missing_pred = model.predict(X_all.loc[mask_na])
    # 평점 범위(0~5) 클리핑
    y_missing_pred = np.clip(y_missing_pred, 0, 5)
    df.loc[mask_na, target] = y_missing_pred
    print(f"결측 대체 완료: {mask_na.sum()}개 행을 채웠습니다.")
else:
    print("대체할 review_scores_rating 결측치가 없습니다.")

# 8) 결과 저장 ----------------------------------------------------------------
out_path = "/content/combined_dataset_filled.csv"
df.to_csv(out_path, index=False)
print(f"저장 완료: {out_path}")

# (선택) 중요 피처 대략 확인 - 트리 내장 중요도(전처리 이후 피처 이름 추적)
try:
    feature_names = []
    # med_ind: 원래 컬럼 수 + 동일한 수의 결측 인디케이터가 뒤에 붙음
    # 이름을 명확히 하기 위해 수동으로 구성
    if median_with_indicator_cols:
        feature_names.extend(median_with_indicator_cols)
        feature_names.extend([f"{c}_is_missing" for c in median_with_indicator_cols])
    feature_names.extend(zero_fill_cols)
    importances = model.named_steps["rf"].feature_importances_
    fi = pd.DataFrame({"feature": feature_names, "importance": importances})\
           .sort_values("importance", ascending=False)
    print("\n[FAST] Feature importance (top 15)\n", fi.head(15).to_string(index=False))
except Exception as e:
    print("[INFO] 중요도 출력 스킵:", e)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("/content/combined_dataset.csv")

# host_is_superhost 0/1로 변환
df["host_is_superhost"] = (
    df["host_is_superhost"]
      .astype(str).str.lower().str.strip()
      .map({"t":1,"true":1,"y":1,"yes":1,"1":1,"f":0,"false":0,"n":0,"no":0,"0":0})
)

df = df[df["host_is_superhost"].notna()]
df = df[df["review_scores_rating"].notna()]

super_hosts = df[df["host_is_superhost"]==1]["review_scores_rating"]
normal_hosts = df[df["host_is_superhost"]==0]["review_scores_rating"]

# Welch t-test (양측)
t_stat, p_two_sided = ttest_ind(super_hosts, normal_hosts, equal_var=False)

# 단측 p-value (슈퍼호스트가 높다는 방향만 검증)
p_one_sided = p_two_sided / 2 if t_stat > 0 else 1 - (p_two_sided / 2)

print(f"슈퍼호스트 평균={super_hosts.mean():.3f}, 일반 평균={normal_hosts.mean():.3f}")
print(f"t={t_stat:.3f}, 양측 p={p_two_sided:.5f}, 단측 p={p_one_sided:.5f}")
```

```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import statsmodels.formula.api as smf

# ======================== 0) Load ========================
df = pd.read_csv("/content/combined_dataset.csv")

# ======================== 1) Feature prep ========================
# amenity_count (쉼표 기준 개수)
if "amenities" in df.columns:
    df["amenity_count"] = (
        df["amenities"].fillna("").astype(str)
          .apply(lambda s: len([x for x in s.split(",") if x.strip()]))
    )

# host_response_rate → 숫자 (예: "95%" → 95.0)
if "host_response_rate" in df.columns:
    df["host_response_rate_num"] = (
        df["host_response_rate"].astype(str).str.replace("%", "", regex=False)
    )
    df["host_response_rate_num"] = pd.to_numeric(df["host_response_rate_num"], errors="coerce")

# price → 숫자
if "price" in df.columns:
    df["price"] = (
        df["price"].astype(str)
          .str.replace("$","", regex=False)
          .str.replace(",","", regex=False)
    )
    df["price"] = pd.to_numeric(df["price"], errors="coerce")

# 타깃(결과) 숫자 변환
targets = {
    "rating": "review_scores_rating",
    "comm": "review_scores_communication",
    "value": "review_scores_value",
    "rpm": "reviews_per_month",
}
for c in targets.values():
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ======================== 2) One-sided t-tests ========================
def one_sided_t(a, b):
    """H1: mean(a) > mean(b) 단측 p-value"""
    a = pd.to_numeric(a, errors="coerce").dropna()
    b = pd.to_numeric(b, errors="coerce").dropna()
    if len(a) < 2 or len(b) < 2:
        return np.nan, np.nan, a.mean() if len(a) else np.nan, b.mean() if len(b) else np.nan, len(a), len(b)
    t2, p2 = ttest_ind(a, b, equal_var=False)
    p1 = p2/2 if t2 > 0 else 1 - p2/2
    return t2, p1, a.mean(), b.mean(), len(a), len(b)

cut = 80  # 응답률 컷(%) — 필요 시 70/90 등으로 바꿔가며 테스트 가능
hi = df[df.get("host_response_rate_num", pd.Series(index=df.index)) >= cut]
lo = df[df.get("host_response_rate_num", pd.Series(index=df.index)) <  cut]

print(f"\n=== One-sided t-tests (High >= {cut}% vs Low) ===")
for name, col in targets.items():
    if col in df.columns and "host_response_rate_num" in df.columns:
        t, p, m_hi, m_lo, n_hi, n_lo = one_sided_t(hi[col], lo[col])
        print(f"[{name}] n_hi={n_hi}, n_lo={n_lo} | "
              f"mean_hi={m_hi:.3f}, mean_lo={m_lo:.3f} | t={t:.2f}, one-sided p={p:.5f}")

# ======================== 3) OLS with controls (imputed) ========================
# 사용할 컬럼 정의
num_cols = ["host_response_rate_num", "review_scores_rating", "price", "amenity_count", "availability_365"]
cat_cols = []
if "room_type" in df.columns: cat_cols.append("room_type")
if "neighbourhood_group_cleansed" in df.columns: cat_cols.append("neighbourhood_group_cleansed")

use_cols = [c for c in num_cols + cat_cols if c in df.columns]
reg_df = df[use_cols].copy()

# 수치형 강제 변환
for c in set(num_cols).intersection(reg_df.columns):
    reg_df[c] = pd.to_numeric(reg_df[c], errors="coerce")

# 결측 임퓨트: 수치형=중앙값, 범주형='Unknown'
for c in set(num_cols).intersection(reg_df.columns):
    med = reg_df[c].median(skipna=True)
    reg_df[c] = reg_df[c].fillna(med)

for c in cat_cols:
    reg_df[c] = reg_df[c].astype("category").cat.add_categories(["Unknown"]).fillna("Unknown")

# 변동성 체크
if "host_response_rate_num" not in reg_df.columns or reg_df["host_response_rate_num"].nunique() <= 1:
    raise ValueError("host_response_rate_num이 없거나 변동성이 부족합니다. 원 데이터의 응답률 컬럼을 확인하세요.")

# 회귀식 구성
formula = "review_scores_rating ~ host_response_rate_num"
if "price" in reg_df.columns:          formula += " + price"
if "amenity_count" in reg_df.columns:  formula += " + amenity_count"
if "availability_365" in reg_df.columns: formula += " + availability_365"
if "room_type" in reg_df.columns:      formula += " + C(room_type)"
if "neighbourhood_group_cleansed" in reg_df.columns: formula += " + C(neighbourhood_group_cleansed)"

print(f"\n=== OLS with controls (imputed) ===")
print(f"Formula: {formula}")
print(f"Samples used: {len(reg_df)}")

# 적합 & 요약
model = smf.ols(formula=formula, data=reg_df).fit()
print(model.summary())

# 단측 해석(응답률 계수 > 0)
beta = model.params.get("host_response_rate_num", np.nan)
p_two = model.pvalues.get("host_response_rate_num", np.nan)
if pd.notna(beta) and pd.notna(p_two):
    p_one = p_two/2 if beta > 0 else 1 - p_two/2
    print("\n=== One-sided test for host_response_rate_num (H1: coefficient > 0) ===")
    print(f"coef = {beta:.6f} | two-sided p = {p_two:.6f} | one-sided p = {p_one:.6f}")
else:
    print("\n[WARN] host_response_rate_num 계수를 찾지 못했습니다. 공식/컬럼을 확인하세요.")

```

```python
# run_ssl_pipeline_fixed.py
# 주피터/코랩 호환: parse_known_args 사용 + 함수형 API 제공

import os, json, random, argparse
from pathlib import Path
from typing import Dict, List
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from joblib import dump

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# -------------------- utils --------------------
def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

def parse_price(val):
    if pd.isna(val): return np.nan
    if isinstance(val, (int, float)): return float(val)
    s = str(val)
    for sym in ["$", "€", "₩", ","]:
        s = s.replace(sym, "")
    try: return float(s)
    except: return np.nan

def topk_one_hot(s: pd.Series, k: int, name: str):
    s = s.fillna("UNKNOWN").astype(str)
    top = s.value_counts().nlargest(k).index
    s2 = s.where(s.isin(top), other="OTHER")
    oh = pd.get_dummies(s2, prefix=name)
    return oh, list(top)

def bool_from_any(x):
    return str(x).lower() in {"1","t","true","y","yes"}

# -------------------- dataset --------------------
class ContrastiveListingDataset(Dataset):
    def __init__(self, X: np.ndarray, host_ids: np.ndarray, feature_noise=0.05, mask_prob=0.05):
        self.X = X.astype(np.float32)
        self.noise = feature_noise
        self.mask = mask_prob
        self.host_to_rows: Dict[int, List[int]] = {}
        for i, h in enumerate(host_ids):
            self.host_to_rows.setdefault(int(h), []).append(i)
        self.hosts = list(self.host_to_rows.keys())

    def __len__(self): return len(self.hosts)

    def _augment(self, v: np.ndarray) -> np.ndarray:
        u = v.copy()
        if self.noise > 0: u = u + self.noise * np.random.randn(*u.shape).astype(np.float32)
        if self.mask > 0:
            m = (np.random.rand(*u.shape) < self.mask)
            u[m] = 0.0
        return u.astype(np.float32)

    def __getitem__(self, idx):
        h = self.hosts[idx]
        rows = self.host_to_rows[h]
        if len(rows) >= 2:
            i1, i2 = np.random.choice(rows, size=2, replace=False)
        else:
            i1 = i2 = rows[0]
        return torch.from_numpy(self._augment(self.X[i1])), torch.from_numpy(self._augment(self.X[i2]))

# -------------------- model --------------------
class MLPEncoder(nn.Module):
    def __init__(self, d_in, d_h=512, d_emb=128):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(d_in, d_h), nn.ReLU(inplace=True), nn.Linear(d_h, d_emb))
    def forward(self, x): return self.net(x)

class ProjectionHead(nn.Module):
    def __init__(self, d_in=128, d_h=256, d_out=64):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(d_in, d_h), nn.ReLU(inplace=True), nn.Linear(d_h, d_out))
    def forward(self, h): return F.normalize(self.net(h), dim=1)

# -------------------- training --------------------
def train_contrastive(encoder, projector, dataloader, device, epochs=50, lr=1e-3, temperature=0.5,
                      grad_accum_steps=1, save_every=10, out_dir=Path("./ssl_ckpt")):
    out_dir.mkdir(parents=True, exist_ok=True)
    opt = torch.optim.Adam(list(encoder.parameters()) + list(projector.parameters()), lr=lr)
    for ep in range(1, epochs+1):
        encoder.train(); projector.train()
        running = 0.0
        for step, (x1, x2) in enumerate(dataloader, start=1):
            x1, x2 = x1.to(device), x2.to(device)
            z1, z2 = projector(encoder(x1)), projector(encoder(x2))
            logits = (z1 @ z2.t()) / temperature
            y = torch.arange(logits.size(0), device=device)
            loss = 0.5 * (F.cross_entropy(logits, y) + F.cross_entropy(logits.t(), y))
            (loss/grad_accum_steps).backward()
            if step % grad_accum_steps == 0:
                opt.step(); opt.zero_grad()
            running += loss.item()
        print(f"[Epoch {ep:03d}] loss={running/len(dataloader):.4f}")
        if ep % save_every == 0 or ep == epochs:
            torch.save({"encoder": encoder.state_dict(),
                        "proj": projector.state_dict(),
                        "cfg": {"epochs": epochs, "lr": lr, "temperature": temperature,
                                "grad_accum_steps": grad_accum_steps}},
                       out_dir / f"ssl_epoch{ep}.pt")

# -------------------- pipeline --------------------
def run_ssl_pipeline(
    raw_csv="combined_dataset.csv",
    out_dir="ssl_host_tiers_out",
    epochs=50, batch=512, lr=1e-3, temperature=0.5,
    hidden=512, emb=128, proj_hidden=256, proj_dim=64,
    noise=0.05, mask=0.05, grad_accum=1, save_every=10,
    seed=42, cpu=False, workers=4
):
    set_seed(seed)
    device = torch.device("cuda" if torch.cuda.is_available() and not cpu else "cpu")
    print("Device:", device)

    raw_csv = Path(raw_csv)
    out_dir = Path(out_dir); out_dir.mkdir(parents=True, exist_ok=True)

    # 1) load & clean
    df = pd.read_csv(raw_csv)
    print("Data shape:", df.shape)
    if "price" in df.columns: df["price"] = df["price"].apply(parse_price)
    assert "host_id" in df.columns, "host_id 컬럼이 필요합니다."

    # 2) features
    num_cols_cand = [
        "price","number_of_reviews","reviews_per_month","calculated_host_listings_count",
        "minimum_nights","maximum_nights","availability_30","availability_60",
        "availability_90","availability_365","review_scores_rating",
        "review_scores_accuracy","review_scores_cleanliness","review_scores_checkin",
        "review_scores_communication","review_scores_location","review_scores_value"
    ]
    num_cols = [c for c in num_cols_cand if c in df.columns]
    cat_cols_cand = ["room_type","property_type","neighbourhood_cleansed","bathrooms_text","instant_bookable","host_response_time"]
    cat_cols = [c for c in cat_cols_cand if c in df.columns]

    num_df = df[num_cols].copy().apply(pd.to_numeric, errors="coerce").fillna(df[num_cols].median())
    oh_parts, meta_top = [], {}
    for c in cat_cols:
        k = 30 if c=="property_type" else (50 if c=="neighbourhood_cleansed" else 10)
        oh, top_vals = topk_one_hot(df[c], k=k, name=c)
        oh_parts.append(oh); meta_top[c] = {"k": k, "top_values": top_vals}
    cat_oh = pd.concat(oh_parts, axis=1) if oh_parts else pd.DataFrame(index=df.index)

    X = pd.concat([num_df, cat_oh], axis=1).astype(float)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X.values)
    pd.DataFrame(Xs, columns=X.columns).to_csv(out_dir/"listing_features.csv", index=False)
    dump(scaler, out_dir/"scaler.joblib")
    (out_dir/"feature_metadata.json").write_text(json.dumps({
        "num_cols": num_cols, "cat_cols": cat_cols, "cat_topk": meta_top,
        "feature_columns": X.columns.tolist()
    }, ensure_ascii=False, indent=2))

    host_ids = df["host_id"].values

    # 3) dataloader
    ds = ContrastiveListingDataset(Xs, host_ids, feature_noise=noise, mask_prob=mask)
    dl = DataLoader(ds, batch_size=batch, shuffle=True, num_workers=workers, drop_last=True, pin_memory=(device.type=="cuda"))

    # 4) train
    input_dim = Xs.shape[1]
    encoder = MLPEncoder(input_dim, hidden, emb).to(device)
    projector = ProjectionHead(emb, proj_hidden, proj_dim).to(device)
    ckpt_dir = out_dir / "checkpoints"
    train_contrastive(encoder, projector, dl, device,
                      epochs=epochs, lr=lr, temperature=temperature,
                      grad_accum_steps=grad_accum, save_every=save_every, out_dir=ckpt_dir)

    # 5) embed listings
    encoder.eval(); E=[]
    with torch.no_grad():
        bs = 4096
        for i in range(0, Xs.shape[0], bs):
            xb = torch.from_numpy(Xs[i:i+bs]).to(device).float()
            E.append(encoder(xb).cpu().numpy())
    E = np.vstack(E)

    # 6) host mean
    emb_df = pd.DataFrame(E); emb_df["host_id"] = host_ids
    host_vec = emb_df.groupby("host_id").mean().sort_index()
    host_vec.to_csv(out_dir/"host_vectors_ssl.csv")

    # 7) PCA + KMeans(4)
    pca = PCA(n_components=min(32, host_vec.shape[1]), random_state=seed)
    Z = pca.fit_transform(host_vec.values); dump(pca, out_dir/"pca.joblib")
    km = KMeans(n_clusters=4, n_init=20, random_state=seed).fit(Z)
    labels = km.predict(Z); dump(km, out_dir/"kmeans.joblib")
    tiers = pd.DataFrame({"host_id": host_vec.index, "cluster": labels})

    # 8) tier mapping by superhost rate & avg rating
    super_col = None
    for cand in ["host_is_superhost","is_superhost"]:
        if cand in df.columns: super_col = cand; break
    if super_col:
        tmp = df[["host_id", super_col]].copy()
        tmp[super_col] = tmp[super_col].apply(bool_from_any).astype(int)
        sup_rate = tmp.groupby("host_id")[super_col].mean()
        tiers = tiers.merge(sup_rate.rename("superhost_rate"), on="host_id", how="left")
    else:
        tiers["superhost_rate"] = np.nan
    if "review_scores_rating" in df.columns:
        rating = df.groupby("host_id")["review_scores_rating"].mean()
        tiers = tiers.merge(rating.rename("avg_rating"), on="host_id", how="left")
    else:
        tiers["avg_rating"] = np.nan

    stats = tiers.groupby("cluster")[["superhost_rate","avg_rating"]].mean().fillna(0.0)
    order = stats.sort_values(["superhost_rate","avg_rating"], ascending=[True,True]).index.tolist()
    names = ["Rookie","Rising","Super","Master"]
    tiers["tier"] = tiers["cluster"].map({c:names[i] for i,c in enumerate(order)})

    tiers.to_csv(out_dir/"host_tiers_ssl.csv", index=False)

    print("\n=== 완료 ===")
    print("출력 폴더:", out_dir.resolve())
    print("  - listing_features.csv")
    print("  - scaler.joblib / feature_metadata.json")
    print("  - checkpoints/ssl_epoch*.pt")
    print("  - host_vectors_ssl.csv")
    print("  - pca.joblib / kmeans.joblib")
    print("  - host_tiers_ssl.csv (최종 결과)")
    print("\n티어 분포:")
    print(tiers["tier"].value_counts())

# -------------------- CLI (parse_known_args로 Colab 호환) --------------------
if __name__ == "__main__":
    ap = argparse.ArgumentParser()
    ap.add_argument("--raw_csv", type=str, default="./combined_dataset.csv")
    ap.add_argument("--out_dir", type=str, default="./ssl_host_tiers_out")
    ap.add_argument("--epochs", type=int, default=50)
    ap.add_argument("--batch", type=int, default=512)
    ap.add_argument("--lr", type=float, default=1e-3)
    ap.add_argument("--temperature", type=float, default=0.5)
    ap.add_argument("--hidden", type=int, default=512)
    ap.add_argument("--emb", type=int, default=128)
    ap.add_argument("--proj_hidden", type=int, default=256)
    ap.add_argument("--proj_dim", type=int, default=64)
    ap.add_argument("--noise", type=float, default=0.05)
    ap.add_argument("--mask", type=float, default=0.05)
    ap.add_argument("--grad_accum", type=int, default=1)
    ap.add_argument("--save_every", type=int, default=10)
    ap.add_argument("--seed", type=int, default=42)
    ap.add_argument("--cpu", action="store_true")
    ap.add_argument("--workers", type=int, default=4)

    args, _unknown = ap.parse_known_args()  # <-- 핵심: 알 수 없는 인자는 무시
    run_ssl_pipeline(**vars(args))
```

```python
# ranked_tiering_ssl_fixed.py
import argparse, json
from pathlib import Path
import numpy as np, pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, Ridge

def parse_price(v):
    if pd.isna(v): return np.nan
    if isinstance(v,(int,float)): return float(v)
    s=str(v)
    for sym in ["$", "€", "₩", ","]: s=s.replace(sym,"")
    try: return float(s)
    except: return np.nan

def zscore(s):
    s = s.astype(float)
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def safe_log1p(s):
    s = pd.to_numeric(s, errors="coerce").fillna(0.0)
    return np.log1p(np.clip(s, 0, None))

def bool_from_any(x):
    return str(x).lower() in {"1","t","true","y","yes"}

def build_host_kpis(df: pd.DataFrame) -> pd.DataFrame:
    agg_defs = {
        "avg_rating": ("review_scores_rating", "mean", True),
        "sum_reviews": ("number_of_reviews", "sum", True),
        "avg_reviews_per_month": ("reviews_per_month", "mean", True),
        "median_price": ("price", "median", False),
        "availability_30": ("availability_30", "mean", False),
        "availability_90": ("availability_90", "mean", False),
        "availability_365": ("availability_365", "mean", False),
    }
    if "price" in df.columns:
        df = df.copy()
        df["price"] = df["price"].apply(parse_price)

    out = pd.DataFrame({"host_id": df["host_id"]}).drop_duplicates().set_index("host_id")
    for name, (col, how, _) in agg_defs.items():
        if col not in df.columns: continue
        if how == "mean":
            s = df.groupby("host_id")[col].mean()
        elif how == "sum":
            s = df.groupby("host_id")[col].sum()
        elif how == "median":
            s = df.groupby("host_id")[col].median()
        else:
            continue
        out[name] = s
    return out

def build_score(host_kpis: pd.DataFrame,
                ssl_host_vec: pd.DataFrame|None,
                weak_labels: pd.Series|None,
                weights: dict|None=None,
                alpha_ssl=0.5) -> pd.Series:
    dir_map = {
        "avg_rating": +1, "sum_reviews": +1, "avg_reviews_per_month": +1,
        "median_price": -1, "availability_30": -1, "availability_90": -1, "availability_365": -1,
    }
    use_cols = [c for c in host_kpis.columns if c in dir_map]
    if not use_cols:
        raise ValueError("KPI가 없습니다. 평점/리뷰/가용일수 등 열이 데이터에 있는지 확인하세요.")

    Z = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        s = host_kpis[c]
        if c in {"sum_reviews"}:
            s = safe_log1p(s)
        z = zscore(pd.to_numeric(s, errors="coerce").fillna(s.median()))
        Z[c] = dir_map[c] * z

    if weights is None:
        weights = {c: 1.0 for c in use_cols}
        if "avg_rating" in weights: weights["avg_rating"] = 2.0

    kpi_score = (Z[list(weights.keys())] * pd.Series(weights)).sum(axis=1)
    kpi_score = (kpi_score - kpi_score.mean()) / (kpi_score.std(ddof=0) + 1e-9)

    final = kpi_score.copy()
    if ssl_host_vec is not None and weak_labels is not None and weak_labels.notna().any():
        idx = host_kpis.index.intersection(ssl_host_vec.index).intersection(weak_labels.index)
        if len(idx) > 5:
            X_ssl = ssl_host_vec.loc[idx].select_dtypes(include=[np.number]).values
            y = weak_labels.loc[idx].apply(bool_from_any).astype(int).values
            scaler = StandardScaler(); Xs = scaler.fit_transform(X_ssl)
            try:
                clf = LogisticRegression(max_iter=1000); clf.fit(Xs, y)
                s = clf.predict_proba(Xs)[:,1]
            except Exception:
                reg = Ridge(alpha=1.0); reg.fit(Xs, y)
                s = reg.predict(Xs)
            s = (s - s.mean()) / (s.std() + 1e-9)
            ssl_score = pd.Series(s, index=idx)
            final.loc[idx] = alpha_ssl*kpi_score.loc[idx] + (1-alpha_ssl)*ssl_score
    return final.rename("score")

def cut_into_tiers(score: pd.Series, p_master=0.05, p_super=0.15, p_rising=0.30):
    s = score.sort_values(ascending=False)
    n = len(s)
    n_master = max(1, int(round(p_master*n)))
    n_super  = max(1, int(round(p_super*n)))
    n_rising = max(1, int(round(p_rising*n)))
    idx_master = s.index[:n_master]
    idx_super  = s.index[n_master:n_master+n_super]
    idx_rising = s.index[n_master+n_super:n_master+n_super+n_rising]
    idx_rookie = s.index[n_master+n_super+n_rising:]
    tier = pd.Series(index=s.index, dtype=object)
    tier.loc[idx_master] = "Master"
    tier.loc[idx_super]  = "Super"
    tier.loc[idx_rising] = "Rising"
    tier.loc[idx_rookie] = "Rookie"
    return tier

def tier_report(df_host, tiers,
                kpi_cols=("avg_rating","sum_reviews","avg_reviews_per_month","median_price","availability_30","availability_90","availability_365")):
    rep = df_host.copy(); rep["tier"] = tiers
    g = rep.groupby("tier")[list(c for c in kpi_cols if c in rep.columns)].mean()
    order = ["Rookie","Rising","Super","Master"]; g = g.reindex(order)
    dir_map = {"avg_rating":"+","sum_reviews":"+","avg_reviews_per_month":"+",
               "median_price":"-","availability_30":"-","availability_90":"-","availability_365":"-"}
    violations = []
    for c in g.columns:
        d = dir_map.get(c, "+"); vals = g[c].values
        if np.isnan(vals).any(): continue
        if d == "+" and not np.all(np.diff(vals) >= -1e-9): violations.append(f"{c}: 기대 ↑ 단조 위반")
        if d == "-" and not np.all(np.diff(vals) <=  1e-9): violations.append(f"{c}: 기대 ↓ 단조 위반")
    return g, violations

def run_ranked_tiering(raw_csv="combined_dataset.csv",
                       ssl_host_vec="",
                       out_dir="ranked_tiers_out",
                       p_master=0.05, p_super=0.15, p_rising=0.30,
                       alpha_ssl=0.5):
    out_dir = Path(out_dir); out_dir.mkdir(parents=True, exist_ok=True)
    df = pd.read_csv(raw_csv)
    assert "host_id" in df.columns, "host_id 컬럼 필요"

    host_kpis = build_host_kpis(df)

    weak_labels = None
    for cand in ["host_is_superhost","is_superhost"]:
        if cand in df.columns:
            weak_labels = df.groupby("host_id")[cand].first()
            break

    ssl_df = None
    if ssl_host_vec and Path(ssl_host_vec).exists():
        ssl_df = pd.read_csv(ssl_host_vec, index_col=0).select_dtypes(include=[np.number])

    score = build_score(host_kpis, ssl_df, weak_labels, alpha_ssl=alpha_ssl)
    tiers = cut_into_tiers(score, p_master=p_master, p_super=p_super, p_rising=p_rising)

    out = pd.DataFrame({"host_id": score.index, "score": score.values, "tier": tiers.loc[score.index].values})
    out.to_csv(out_dir/"host_tiers_ranked.csv", index=False)

    report, violations = tier_report(host_kpis, tiers)
    report.to_csv(out_dir/"tier_kpi_report.csv")

    meta = {"p_master": p_master, "p_super": p_super, "p_rising": p_rising,
            "alpha_ssl": alpha_ssl, "used_ssl": ssl_df is not None,
            "used_weak_label": weak_labels is not None}
    (out_dir/"ranked_tiering_meta.json").write_text(json.dumps(meta, ensure_ascii=False, indent=2))

    print("등급 분포:\n", out["tier"].value_counts(), "\n")
    print("등급별 KPI 평균:\n", report, "\n")
    if violations:
        print("⚠ 단조성 위반:", *violations, sep="\n - ")
    else:
        print("✅ 주요 KPI 단조성 만족")
    print("완료! 결과:", (out_dir/"host_tiers_ranked.csv").resolve())

# ---------------- CLI(주피터 호환) ----------------
if __name__ == "__main__":
    ap = argparse.ArgumentParser()
    ap.add_argument("--raw_csv", default="./combined_dataset.csv")
    ap.add_argument("--ssl_host_vec", default="")
    ap.add_argument("--out_dir", default="./ranked_tiers_out")
    ap.add_argument("--p_master", type=float, default=0.05)
    ap.add_argument("--p_super", type=float, default=0.15)
    ap.add_argument("--p_rising", type=float, default=0.30)
    ap.add_argument("--alpha_ssl", type=float, default=0.5)
    args, _ = ap.parse_known_args()   # <- 주피터가 끼워넣는 -f 인자 무시
    run_ranked_tiering(**vars(args))
```

```python
# -*- coding: utf-8 -*-
"""
Self-Supervised(Contrastive) + Ranked Tiering (Quantile Cut) End-to-End
- 주피터/코랩 호환: parse_known_args 사용
- 함수형 API: run_ssl_ranked(...)

사용 예:
  (A) 노트북 셀에서 함수 호출:
    from run_ssl_ranked_end2end import run_ssl_ranked
    run_ssl_ranked(raw_csv="combined_dataset.csv", out_dir="out", epochs=30)

  (B) 스크립트 실행:
    !python run_ssl_ranked_end2end.py --raw_csv combined_dataset.csv --out_dir out --epochs 30
"""

import os, json, random, argparse
from pathlib import Path
from typing import Dict, List, Tuple, Optional

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression, Ridge
from joblib import dump

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader


# -------------------- utils --------------------
def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

def parse_price(val):
    if pd.isna(val): return np.nan
    if isinstance(val, (int, float)): return float(val)
    s = str(val)
    for sym in ["$", "€", "₩", ","]:
        s = s.replace(sym, "")
    try: return float(s)
    except: return np.nan

def topk_one_hot(s: pd.Series, k: int, name: str):
    s = s.fillna("UNKNOWN").astype(str)
    top = s.value_counts().nlargest(k).index
    s2 = s.where(s.isin(top), other="OTHER")
    oh = pd.get_dummies(s2, prefix=name)
    return oh, list(top)

def bool_from_any(x):
    return str(x).lower() in {"1","t","true","y","yes"}

def zscore(s):
    s = pd.to_numeric(s, errors="coerce").astype(float)
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def safe_log1p(s):
    s = pd.to_numeric(s, errors="coerce").fillna(0.0)
    return np.log1p(np.clip(s, 0, None))


# -------------------- dataset (contrastive SSL) --------------------
class ContrastiveListingDataset(Dataset):
    """
    host_id를 기준으로 같은 호스트의 두 리스팅(또는 같은 리스팅의 두 증강뷰)을 positive pair로 구성
    """
    def __init__(self, X: np.ndarray, host_ids: np.ndarray,
                 feature_noise=0.05, mask_prob=0.05):
        self.X = X.astype(np.float32)
        self.noise = feature_noise
        self.mask = mask_prob
        self.host_to_rows: Dict[int, List[int]] = {}
        for i, h in enumerate(host_ids):
            self.host_to_rows.setdefault(int(h), []).append(i)
        self.hosts = list(self.host_to_rows.keys())

    def __len__(self):
        return len(self.hosts)

    def _augment(self, v: np.ndarray) -> np.ndarray:
        u = v.copy()
        if self.noise > 0:
            u = u + self.noise * np.random.randn(*u.shape).astype(np.float32)
        if self.mask > 0:
            m = (np.random.rand(*u.shape) < self.mask)
            u[m] = 0.0
        return u.astype(np.float32)

    def __getitem__(self, idx):
        h = self.hosts[idx]
        rows = self.host_to_rows[h]
        if len(rows) >= 2:
            i1, i2 = np.random.choice(rows, size=2, replace=False)
        else:
            i1 = i2 = rows[0]
        return torch.from_numpy(self._augment(self.X[i1])), torch.from_numpy(self._augment(self.X[i2]))


# -------------------- models --------------------
class MLPEncoder(nn.Module):
    def __init__(self, d_in, d_h=512, d_emb=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_emb)
        )
    def forward(self, x):
        return self.net(x)

class ProjectionHead(nn.Module):
    def __init__(self, d_in=128, d_h=256, d_out=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_out)
        )
    def forward(self, h):
        z = self.net(h)
        return F.normalize(z, dim=1)  # 코사인 유사도 안정화


# -------------------- training (contrastive InfoNCE) --------------------
def train_contrastive(encoder, projector, dataloader, device,
                      epochs=50, lr=1e-3, temperature=0.5,
                      grad_accum_steps=1, save_every=10, out_dir=Path("./ssl_ckpt")):
    out_dir.mkdir(parents=True, exist_ok=True)
    opt = torch.optim.Adam(list(encoder.parameters()) + list(projector.parameters()), lr=lr)
    for ep in range(1, epochs+1):
        encoder.train(); projector.train()
        running = 0.0
        for step, (x1, x2) in enumerate(dataloader, start=1):
            x1, x2 = x1.to(device), x2.to(device)
            z1, z2 = projector(encoder(x1)), projector(encoder(x2))
            logits = (z1 @ z2.t()) / temperature             # (B, B)
            y = torch.arange(logits.size(0), device=device) # 0..B-1 (i번째는 i번째와 매칭)
            loss = 0.5 * (F.cross_entropy(logits, y) + F.cross_entropy(logits.t(), y))  # 양방향 InfoNCE
            (loss/grad_accum_steps).backward()
            if step % grad_accum_steps == 0:
                opt.step(); opt.zero_grad()
            running += loss.item()
        print(f"[Epoch {ep:03d}] loss={running/len(dataloader):.4f}")
        if ep % save_every == 0 or ep == epochs:
            torch.save({
                "encoder": encoder.state_dict(),
                "proj": projector.state_dict(),
                "cfg": {"epochs": epochs, "lr": lr, "temperature": temperature,
                        "grad_accum_steps": grad_accum_steps}
            }, out_dir / f"ssl_epoch{ep}.pt")


# -------------------- KPI 집계 --------------------
def build_host_kpis(df: pd.DataFrame) -> pd.DataFrame:
    agg_defs = {
        "avg_rating": ("review_scores_rating", "mean", True),
        "sum_reviews": ("number_of_reviews", "sum", True),
        "avg_reviews_per_month": ("reviews_per_month", "mean", True),
        "median_price": ("price", "median", False),
        "availability_30": ("availability_30", "mean", False),
        "availability_90": ("availability_90", "mean", False),
        "availability_365": ("availability_365", "mean", False),
    }
    if "price" in df.columns:
        df = df.copy()
        df["price"] = df["price"].apply(parse_price)

    out = pd.DataFrame({"host_id": df["host_id"]}).drop_duplicates().set_index("host_id")
    for name, (col, how, _) in agg_defs.items():
        if col not in df.columns: continue
        if how == "mean":
            s = df.groupby("host_id")[col].mean()
        elif how == "sum":
            s = df.groupby("host_id")[col].sum()
        elif how == "median":
            s = df.groupby("host_id")[col].median()
        else:
            continue
        out[name] = s
    return out


# -------------------- 점수 산출 (KPI + SSL 약지도 블렌딩) --------------------
def build_score(host_kpis: pd.DataFrame,
                ssl_host_vec: Optional[pd.DataFrame],
                weak_labels: Optional[pd.Series],
                weights: Optional[dict]=None,
                alpha_ssl=0.5) -> pd.Series:
    dir_map = {
        "avg_rating": +1, "sum_reviews": +1, "avg_reviews_per_month": +1,
        "median_price": -1, "availability_30": -1, "availability_90": -1, "availability_365": -1,
    }
    use_cols = [c for c in host_kpis.columns if c in dir_map]
    if not use_cols:
        raise ValueError("KPI가 없습니다. 평점/리뷰/가용일수 등 열이 데이터에 있는지 확인하세요.")

    Z = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        s = host_kpis[c]
        if c in {"sum_reviews"}:
            s = safe_log1p(s)
        z = zscore(pd.to_numeric(s, errors="coerce").fillna(s.median()))
        Z[c] = dir_map[c] * z

    if weights is None:
        weights = {c: 1.0 for c in use_cols}
        if "avg_rating" in weights: weights["avg_rating"] = 2.0

    kpi_score = (Z[list(weights.keys())] * pd.Series(weights)).sum(axis=1)
    kpi_score = (kpi_score - kpi_score.mean()) / (kpi_score.std(ddof=0) + 1e-9)

    final = kpi_score.copy()

    # 선택: SSL 임베딩 + 약한 라벨(슈퍼호스트)로 세미지도 점수 보강
    if ssl_host_vec is not None and weak_labels is not None and weak_labels.notna().any():
        idx = host_kpis.index.intersection(ssl_host_vec.index).intersection(weak_labels.index)
        if len(idx) > 5:
            X_ssl = ssl_host_vec.loc[idx].select_dtypes(include=[np.number]).values
            y = weak_labels.loc[idx].apply(bool_from_any).astype(int).values
            scaler = StandardScaler(); Xs = scaler.fit_transform(X_ssl)
            try:
                clf = LogisticRegression(max_iter=1000); clf.fit(Xs, y)
                s = clf.predict_proba(Xs)[:,1]
            except Exception:
                reg = Ridge(alpha=1.0); reg.fit(Xs, y)
                s = reg.predict(Xs)
            s = (s - s.mean()) / (s.std() + 1e-9)
            ssl_score = pd.Series(s, index=idx)
            final.loc[idx] = alpha_ssl*kpi_score.loc[idx] + (1-alpha_ssl)*ssl_score

    return final.rename("score")


# -------------------- 퍼센타일 등급 컷 --------------------
def cut_into_tiers(score: pd.Series, p_master=0.05, p_super=0.15, p_rising=0.30):
    s = score.sort_values(ascending=False)
    n = len(s)
    n_master = max(1, int(round(p_master*n)))
    n_super  = max(1, int(round(p_super*n)))
    n_rising = max(1, int(round(p_rising*n)))
    idx_master = s.index[:n_master]
    idx_super  = s.index[n_master:n_master+n_super]
    idx_rising = s.index[n_master+n_super:n_master+n_super+n_rising]
    idx_rookie = s.index[n_master+n_super+n_rising:]
    tier = pd.Series(index=s.index, dtype=object)
    tier.loc[idx_master] = "Master"
    tier.loc[idx_super]  = "Super"
    tier.loc[idx_rising] = "Rising"
    tier.loc[idx_rookie] = "Rookie"
    return tier


# -------------------- 등급 리포트 & 단조성 체크 --------------------
def tier_report(df_host, tiers,
                kpi_cols=("avg_rating","sum_reviews","avg_reviews_per_month","median_price","availability_30","availability_90","availability_365")):
    rep = df_host.copy(); rep["tier"] = tiers
    g = rep.groupby("tier")[list(c for c in kpi_cols if c in rep.columns)].mean()
    order = ["Rookie","Rising","Super","Master"]; g = g.reindex(order)
    dir_map = {"avg_rating":"+","sum_reviews":"+","avg_reviews_per_month":"+",
               "median_price":"-","availability_30":"-","availability_90":"-","availability_365":"-"}
    violations = []
    for c in g.columns:
        d = dir_map.get(c, "+"); vals = g[c].values
        if np.isnan(vals).any(): continue
        if d == "+" and not np.all(np.diff(vals) >= -1e-9):
            violations.append(f"{c}: 기대 ↑ 단조 위반")
        if d == "-" and not np.all(np.diff(vals) <=  1e-9):
            violations.append(f"{c}: 기대 ↓ 단조 위반")
    return g, violations


# -------------------- End-to-End 파이프라인 --------------------
def run_ssl_ranked(
    raw_csv="combined_dataset.csv",
    out_dir="out",
    # SSL 학습
    epochs=50, batch=512, lr=1e-3, temperature=0.5,
    hidden=512, emb=128, proj_hidden=256, proj_dim=64,
    noise=0.05, mask=0.05, grad_accum=1, save_every=10,
    # 등급 컷
    p_master=0.05, p_super=0.15, p_rising=0.30,
    alpha_ssl=0.5,
    # 실행 옵션
    seed=42, cpu=False, workers=4
):
    set_seed(seed)
    device = torch.device("cuda" if torch.cuda.is_available() and not cpu else "cpu")
    print("Device:", device)

    raw_csv = Path(raw_csv)
    out_dir = Path(out_dir); out_dir.mkdir(parents=True, exist_ok=True)
    ssl_dir = out_dir / "ssl"
    ranked_dir = out_dir / "ranked_tiers"
    ssl_dir.mkdir(exist_ok=True, parents=True)
    ranked_dir.mkdir(exist_ok=True, parents=True)

    # 1) 데이터 로드 & 기본 정리
    df = pd.read_csv(raw_csv)
    print("Data shape:", df.shape)
    if "price" in df.columns: df["price"] = df["price"].apply(parse_price)
    assert "host_id" in df.columns, "host_id 컬럼이 필요합니다."

    # 2) 피처 엔지니어링(리스팅 단위)
    num_cols_cand = [
        "price","number_of_reviews","reviews_per_month","calculated_host_listings_count",
        "minimum_nights","maximum_nights","availability_30","availability_60",
        "availability_90","availability_365","review_scores_rating",
        "review_scores_accuracy","review_scores_cleanliness","review_scores_checkin",
        "review_scores_communication","review_scores_location","review_scores_value"
    ]
    num_cols = [c for c in num_cols_cand if c in df.columns]

    cat_cols_cand = ["room_type","property_type","neighbourhood_cleansed",
                     "bathrooms_text","instant_bookable","host_response_time"]
    cat_cols = [c for c in cat_cols_cand if c in df.columns]

    num_df = df[num_cols].copy().apply(pd.to_numeric, errors="coerce")
    num_df = num_df.fillna(num_df.median())

    oh_parts, meta_top = [], {}
    for c in cat_cols:
        k = 30 if c == "property_type" else (50 if c == "neighbourhood_cleansed" else 10)
        oh, top_vals = topk_one_hot(df[c], k=k, name=c)
        oh_parts.append(oh); meta_top[c] = {"k": k, "top_values": top_vals}
    cat_oh = pd.concat(oh_parts, axis=1) if oh_parts else pd.DataFrame(index=df.index)

    X = pd.concat([num_df, cat_oh], axis=1).astype(float)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X.values)
    pd.DataFrame(Xs, columns=X.columns).to_csv(ssl_dir/"listing_features.csv", index=False)
    dump(scaler, ssl_dir/"scaler.joblib")
    (ssl_dir/"feature_metadata.json").write_text(json.dumps({
        "num_cols": num_cols, "cat_cols": cat_cols, "cat_topk": meta_top,
        "feature_columns": X.columns.tolist()
    }, ensure_ascii=False, indent=2))

    host_ids = df["host_id"].values

    # 3) 대조학습 데이터로더
    ds = ContrastiveListingDataset(Xs, host_ids, feature_noise=noise, mask_prob=mask)
    dl = DataLoader(ds, batch_size=batch, shuffle=True, num_workers=workers,
                    drop_last=True, pin_memory=(device.type=="cuda"))

    # 4) 인코더/프로젝션 학습(SSL)
    input_dim = Xs.shape[1]
    encoder = MLPEncoder(input_dim, hidden, emb).to(device)
    projector = ProjectionHead(emb, proj_hidden, proj_dim).to(device)
    ckpt_dir = ssl_dir / "checkpoints"
    train_contrastive(encoder, projector, dl, device,
                      epochs=epochs, lr=lr, temperature=temperature,
                      grad_accum_steps=grad_accum, save_every=save_every, out_dir=ckpt_dir)

    # 5) 리스팅 임베딩 → 호스트 평균 임베딩
    encoder.eval(); E=[]
    with torch.no_grad():
        bs = 4096
        for i in range(0, Xs.shape[0], bs):
            xb = torch.from_numpy(Xs[i:i+bs]).to(device).float()
            E.append(encoder(xb).cpu().numpy())
    E = np.vstack(E)

    emb_df = pd.DataFrame(E); emb_df["host_id"] = host_ids
    host_vec = emb_df.groupby("host_id").mean().sort_index()  # (n_hosts, emb_dim)
    host_vec.to_csv(ssl_dir/"host_vectors_ssl.csv")

    # 6) KPI 집계
    host_kpis = build_host_kpis(df)

    # 7) 약한 라벨(슈퍼호스트) 추출(있으면)
    weak_labels = None
    for cand in ["host_is_superhost","is_superhost"]:
        if cand in df.columns:
            weak_labels = df.groupby("host_id")[cand].first()
            break

    # 8) KPI+SSL 점수 → 퍼센타일 등급
    score = build_score(host_kpis=host_kpis,
                        ssl_host_vec=host_vec,
                        weak_labels=weak_labels,
                        alpha_ssl=alpha_ssl)
    tiers = cut_into_tiers(score, p_master=p_master, p_super=p_super, p_rising=p_rising)

    # 9) 결과 저장
    out = pd.DataFrame({"host_id": score.index, "score": score.values, "tier": tiers.loc[score.index].values})
    out.to_csv(ranked_dir/"host_tiers_ranked.csv", index=False)

    report, violations = tier_report(host_kpis, tiers)
    report.to_csv(ranked_dir/"tier_kpi_report.csv")

    meta = {
        "epochs": epochs, "batch": batch, "lr": lr, "temperature": temperature,
        "hidden": hidden, "emb": emb, "proj_hidden": proj_hidden, "proj_dim": proj_dim,
        "noise": noise, "mask": mask, "grad_accum": grad_accum, "save_every": save_every,
        "p_master": p_master, "p_super": p_super, "p_rising": p_rising,
        "alpha_ssl": alpha_ssl,
        "used_weak_label": weak_labels is not None
    }
    (ranked_dir/"ranked_tiering_meta.json").write_text(json.dumps(meta, ensure_ascii=False, indent=2))

    print("\n=== 완료 ===")
    print("출력 폴더:", out_dir.resolve())
    print(" [SSL]  ", (ssl_dir/"listing_features.csv").name,
          (ssl_dir/"scaler.joblib").name, (ssl_dir/"feature_metadata.json").name, sep="\n   - ")
    print("   - checkpoints/ssl_epoch*.pt")
    print("   - host_vectors_ssl.csv")
    print(" [Tier] ", (ranked_dir/"host_tiers_ranked.csv").name,
          (ranked_dir/"tier_kpi_report.csv").name,
          (ranked_dir/"ranked_tiering_meta.json").name, sep="\n   - ")
    print("\n티어 분포:")
    print(out["tier"].value_counts())
    if violations:
        print("\n⚠ 단조성 위반:")
        for v in violations: print(" -", v)
    else:
        print("\n✅ 주요 KPI 단조성 만족")


# -------------------- CLI(주피터/코랩 호환) --------------------
if __name__ == "__main__":
    ap = argparse.ArgumentParser()
    ap.add_argument("--raw_csv", type=str, default="./combined_dataset.csv")
    ap.add_argument("--out_dir", type=str, default="./out")
    # SSL
    ap.add_argument("--epochs", type=int, default=50)
    ap.add_argument("--batch", type=int, default=512)
    ap.add_argument("--lr", type=float, default=1e-3)
    ap.add_argument("--temperature", type=float, default=0.5)
    ap.add_argument("--hidden", type=int, default=512)
    ap.add_argument("--emb", type=int, default=128)
    ap.add_argument("--proj_hidden", type=int, default=256)
    ap.add_argument("--proj_dim", type=int, default=64)
    ap.add_argument("--noise", type=float, default=0.05)
    ap.add_argument("--mask", type=float, default=0.05)
    ap.add_argument("--grad_accum", type=int, default=1)
    ap.add_argument("--save_every", type=int, default=10)
    # Ranked tier cut
    ap.add_argument("--p_master", type=float, default=0.05)
    ap.add_argument("--p_super", type=float, default=0.15)
    ap.add_argument("--p_rising", type=float, default=0.30)
    ap.add_argument("--alpha_ssl", type=float, default=0.5)
    # Exec
    ap.add_argument("--seed", type=int, default=42)
    ap.add_argument("--cpu", action="store_true")
    ap.add_argument("--workers", type=int, default=4)

    args, _ = ap.parse_known_args()  # 주피터가 끼워넣는 -f 인자 무시
    run_ssl_ranked(**vars(args))
```

```python
# -*- coding: utf-8 -*-
"""
End-to-End: SSL(대조학습) + KPI/SSL 점수화 + 퍼센타일 4등급 + '증거 패키지' 생성
- 주피터/코랩 호환(parse_known_args)
- 함수형 API: run_ssl_ranked_with_evidence(...)

산출물:
  out/
    ssl/
      listing_features.csv, scaler.joblib, feature_metadata.json
      checkpoints/ssl_epoch*.pt
      host_vectors_ssl.csv
    ranked_tiers/
      host_tiers_ranked.csv
      tier_kpi_report.csv
      score_cutoffs.json
      kpi_contributions.csv
      tier_distribution.json
      evidence_report.md
"""

import os, json, random, argparse
from pathlib import Path
from typing import Dict, List, Optional, Tuple

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression, Ridge
from joblib import dump

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader


# -------------------- utils --------------------
def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

def parse_price(val):
    if pd.isna(val): return np.nan
    if isinstance(val, (int, float)): return float(val)
    s = str(val)
    for sym in ["$", "€", "₩", ","]:
        s = s.replace(sym, "")
    try: return float(s)
    except: return np.nan

def topk_one_hot(s: pd.Series, k: int, name: str):
    s = s.fillna("UNKNOWN").astype(str)
    top = s.value_counts().nlargest(k).index
    s2 = s.where(s.isin(top), other="OTHER")
    oh = pd.get_dummies(s2, prefix=name)
    return oh, list(top)

def bool_from_any(x):
    return str(x).lower() in {"1","t","true","y","yes"}

def zscore_series(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").astype(float)
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def safe_log1p(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").fillna(0.0)
    return np.log1p(np.clip(s, 0, None))


# -------------------- dataset (contrastive SSL) --------------------
class ContrastiveListingDataset(Dataset):
    """같은 host_id의 두 리스팅(또는 증강뷰)을 positive pair로 구성"""
    def __init__(self, X: np.ndarray, host_ids: np.ndarray,
                 feature_noise=0.05, mask_prob=0.05):
        self.X = X.astype(np.float32)
        self.noise = feature_noise
        self.mask = mask_prob
        self.host_to_rows: Dict[int, List[int]] = {}
        for i, h in enumerate(host_ids):
            self.host_to_rows.setdefault(int(h), []).append(i)
        self.hosts = list(self.host_to_rows.keys())

    def __len__(self): return len(self.hosts)

    def _augment(self, v: np.ndarray) -> np.ndarray:
        u = v.copy()
        if self.noise > 0:
            u = u + self.noise * np.random.randn(*u.shape).astype(np.float32)
        if self.mask > 0:
            m = (np.random.rand(*u.shape) < self.mask)
            u[m] = 0.0
        return u.astype(np.float32)

    def __getitem__(self, idx):
        h = self.hosts[idx]
        rows = self.host_to_rows[h]
        if len(rows) >= 2:
            i1, i2 = np.random.choice(rows, size=2, replace=False)
        else:
            i1 = i2 = rows[0]
        return torch.from_numpy(self._augment(self.X[i1])), torch.from_numpy(self._augment(self.X[i2]))


# -------------------- models --------------------
class MLPEncoder(nn.Module):
    def __init__(self, d_in, d_h=512, d_emb=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_emb)
        )
    def forward(self, x):
        return self.net(x)

class ProjectionHead(nn.Module):
    def __init__(self, d_in=128, d_h=256, d_out=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_out)
        )
    def forward(self, h):
        z = self.net(h)
        return F.normalize(z, dim=1)


# -------------------- training (contrastive InfoNCE) --------------------
def train_contrastive(encoder, projector, dataloader, device,
                      epochs=50, lr=1e-3, temperature=0.5,
                      grad_accum_steps=1, save_every=10, out_dir=Path("./ssl_ckpt")):
    out_dir.mkdir(parents=True, exist_ok=True)
    opt = torch.optim.Adam(list(encoder.parameters()) + list(projector.parameters()), lr=lr)
    for ep in range(1, epochs+1):
        encoder.train(); projector.train()
        running = 0.0
        for step, (x1, x2) in enumerate(dataloader, start=1):
            x1, x2 = x1.to(device), x2.to(device)
            z1, z2 = projector(encoder(x1)), projector(encoder(x2))
            logits = (z1 @ z2.t()) / temperature
            y = torch.arange(logits.size(0), device=device)
            loss = 0.5 * (F.cross_entropy(logits, y) + F.cross_entropy(logits.t(), y))
            (loss/grad_accum_steps).backward()
            if step % grad_accum_steps == 0:
                opt.step(); opt.zero_grad()
            running += loss.item()
        print(f"[Epoch {ep:03d}] loss={running/len(dataloader):.4f}")
        if ep % save_every == 0 or ep == epochs:
            torch.save({
                "encoder": encoder.state_dict(),
                "proj": projector.state_dict(),
                "cfg": {"epochs": epochs, "lr": lr, "temperature": temperature,
                        "grad_accum_steps": grad_accum_steps}
            }, out_dir / f"ssl_epoch{ep}.pt")


# -------------------- KPI 집계 --------------------
def build_host_kpis(df: pd.DataFrame) -> pd.DataFrame:
    agg_defs = {
        "avg_rating": ("review_scores_rating", "mean", True),
        "sum_reviews": ("number_of_reviews", "sum", True),
        "avg_reviews_per_month": ("reviews_per_month", "mean", True),
        "median_price": ("price", "median", False),
        "availability_30": ("availability_30", "mean", False),
        "availability_90": ("availability_90", "mean", False),
        "availability_365": ("availability_365", "mean", False),
    }
    if "price" in df.columns:
        df = df.copy()
        df["price"] = df["price"].apply(parse_price)

    out = pd.DataFrame({"host_id": df["host_id"]}).drop_duplicates().set_index("host_id")
    for name, (col, how, _) in agg_defs.items():
        if col not in df.columns: continue
        if how == "mean":
            s = df.groupby("host_id")[col].mean()
        elif how == "sum":
            s = df.groupby("host_id")[col].sum()
        elif how == "median":
            s = df.groupby("host_id")[col].median()
        else:
            continue
        out[name] = s
    return out


# -------------------- 점수 산출 (KPI + SSL 약지도 블렌딩) --------------------
def build_score_with_components(host_kpis: pd.DataFrame,
                                ssl_host_vec: Optional[pd.DataFrame],
                                weak_labels: Optional[pd.Series],
                                weights: Optional[dict]=None,
                                alpha_ssl=0.5):
    """
    반환:
      score: 최종 점수(Series)
      kpi_score: KPI 점수(Series)
      ssl_score: SSL-약지도 점수(Series 또는 None)
      Z: KPI z-score 테이블(방향 보정 전)
      Z_oriented: 방향/가중 반영된 KPI 기여도 테이블
      meta: dir_map, weights 등 메타
    """
    dir_map = {
        "avg_rating": +1, "sum_reviews": +1, "avg_reviews_per_month": +1,
        "median_price": -1, "availability_30": -1, "availability_90": -1, "availability_365": -1,
    }
    use_cols = [c for c in host_kpis.columns if c in dir_map]
    if not use_cols:
        raise ValueError("KPI가 없습니다. 평점/리뷰/가용일수 등 열이 데이터에 있는지 확인하세요.")

    # KPI z-score
    Z = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        s = host_kpis[c]
        if c in {"sum_reviews"}:
            s = safe_log1p(s)
        z = zscore_series(s.fillna(s.median()))
        Z[c] = z

    # 가중치 (평점 2배 기본)
    if weights is None:
        weights = {c: 1.0 for c in use_cols}
        if "avg_rating" in weights: weights["avg_rating"] = 2.0

    # 방향 보정 + 기여도
    Z_oriented = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        Z_oriented[f"contrib_{c}"] = dir_map[c] * weights[c] * Z[c]

    kpi_score = Z_oriented.sum(axis=1)
    kpi_score = zscore_series(kpi_score)

    # SSL-약지도 점수
    ssl_score = None
    if ssl_host_vec is not None and weak_labels is not None and weak_labels.notna().any():
        idx = host_kpis.index.intersection(ssl_host_vec.index).intersection(weak_labels.index)
        if len(idx) > 5:
            X_ssl = ssl_host_vec.loc[idx].select_dtypes(include=[np.number]).values
            y = weak_labels.loc[idx].apply(bool_from_any).astype(int).values
            scaler = StandardScaler(); Xs = scaler.fit_transform(X_ssl)
            try:
                clf = LogisticRegression(max_iter=1000); clf.fit(Xs, y)
                s = clf.predict_proba(Xs)[:,1]
            except Exception:
                reg = Ridge(alpha=1.0); reg.fit(Xs, y)
                s = reg.predict(Xs)
            s = (s - s.mean()) / (s.std() + 1e-9)
            ssl_score = pd.Series(s, index=idx, name="ssl_score")

    # 최종 점수 블렌딩
    final = kpi_score.copy()
    if ssl_score is not None:
        final.loc[ssl_score.index] = alpha_ssl*kpi_score.loc[ssl_score.index] + (1-alpha_ssl)*ssl_score

    meta = {"dir_map": dir_map, "weights": weights}
    return final.rename("score"), kpi_score.rename("kpi_score"), ssl_score, Z[use_cols], Z_oriented, meta


# -------------------- 퍼센타일 등급 컷 --------------------
def cut_into_tiers_with_cutoffs(score: pd.Series, p_master=0.05, p_super=0.15, p_rising=0.30):
    s = score.sort_values(ascending=False)
    n = len(s)
    n_master = max(1, int(round(p_master*n)))
    n_super  = max(1, int(round(p_super*n)))
    n_rising = max(1, int(round(p_rising*n)))
    idx_master = s.index[:n_master]
    idx_super  = s.index[n_master:n_master+n_super]
    idx_rising = s.index[n_master+n_super:n_master+n_super+n_rising]
    idx_rookie = s.index[n_master+n_super+n_rising:]
    tier = pd.Series(index=s.index, dtype=object)
    tier.loc[idx_master] = "Master"
    tier.loc[idx_super]  = "Super"
    tier.loc[idx_rising] = "Rising"
    tier.loc[idx_rookie] = "Rookie"

    # 커트라인(실제 수치) 계산
    cut_master = s.iloc[n_master-1] if n_master>0 else s.max()
    cut_super  = s.iloc[n_master+n_super-1] if n_super>0 else s.iloc[n_master-1]
    cut_rising = s.iloc[n_master+n_super+n_rising-1] if n_rising>0 else s.iloc[n_master+n_super-1]
    cutoffs = {
        "Master_min_score": float(cut_master),
        "Super_min_score":  float(cut_super),
        "Rising_min_score": float(cut_rising),
    }
    return tier, cutoffs


# -------------------- 등급 리포트 & 단조성 체크 --------------------
def tier_report(df_host, tiers,
                kpi_cols=("avg_rating","sum_reviews","avg_reviews_per_month","median_price","availability_30","availability_90","availability_365")):
    rep = df_host.copy(); rep["tier"] = tiers
    g = rep.groupby("tier")[list(c for c in kpi_cols if c in rep.columns)].mean()
    order = ["Rookie","Rising","Super","Master"]; g = g.reindex(order)
    dir_map = {"avg_rating":"+","sum_reviews":"+","avg_reviews_per_month":"+",
               "median_price":"-","availability_30":"-","availability_90":"-","availability_365":"-"}
    violations = []
    for c in g.columns:
        d = dir_map.get(c, "+"); vals = g[c].values
        if np.isnan(vals).any(): continue
        if d == "+" and not np.all(np.diff(vals) >= -1e-9):
            violations.append(f"{c}: 기대 ↑ 단조 위반")
        if d == "-" and not np.all(np.diff(vals) <=  1e-9):
            violations.append(f"{c}: 기대 ↓ 단조 위반")
    return g, violations


# -------------------- End-to-End + Evidence --------------------
def run_ssl_ranked_with_evidence(
    raw_csv="combined_dataset.csv",
    out_dir="out",
    # SSL 학습
    epochs=50, batch=512, lr=1e-3, temperature=0.5,
    hidden=512, emb=128, proj_hidden=256, proj_dim=64,
    noise=0.05, mask=0.05, grad_accum=1, save_every=10,
    # Ranked tier cut
    p_master=0.05, p_super=0.15, p_rising=0.30, alpha_ssl=0.5,
    # Exec
    seed=42, cpu=False, workers=4
):
    set_seed(seed)
    device = torch.device("cuda" if torch.cuda.is_available() and not cpu else "cpu")
    print("Device:", device)

    raw_csv = Path(raw_csv)
    out_dir = Path(out_dir); out_dir.mkdir(parents=True, exist_ok=True)
    ssl_dir = out_dir / "ssl"
    tier_dir = out_dir / "ranked_tiers"
    ssl_dir.mkdir(exist_ok=True, parents=True)
    tier_dir.mkdir(exist_ok=True, parents=True)

    # 1) 데이터 로드/정리
    df = pd.read_csv(raw_csv)
    print("Data shape:", df.shape)
    if "price" in df.columns: df["price"] = df["price"].apply(parse_price)
    assert "host_id" in df.columns, "host_id 컬럼이 필요합니다."

    # 2) 리스팅 피처 엔지니어링
    num_cols_cand = [
        "price","number_of_reviews","reviews_per_month","calculated_host_listings_count",
        "minimum_nights","maximum_nights","availability_30","availability_60",
        "availability_90","availability_365","review_scores_rating",
        "review_scores_accuracy","review_scores_cleanliness","review_scores_checkin",
        "review_scores_communication","review_scores_location","review_scores_value"
    ]
    num_cols = [c for c in num_cols_cand if c in df.columns]

    cat_cols_cand = ["room_type","property_type","neighbourhood_cleansed",
                     "bathrooms_text","instant_bookable","host_response_time"]
    cat_cols = [c for c in cat_cols_cand if c in df.columns]

    num_df = df[num_cols].copy().apply(pd.to_numeric, errors="coerce")
    num_df = num_df.fillna(num_df.median())

    oh_parts, meta_top = [], {}
    for c in cat_cols:
        k = 30 if c == "property_type" else (50 if c == "neighbourhood_cleansed" else 10)
        oh, top_vals = topk_one_hot(df[c], k=k, name=c)
        oh_parts.append(oh); meta_top[c] = {"k": k, "top_values": top_vals}
    cat_oh = pd.concat(oh_parts, axis=1) if oh_parts else pd.DataFrame(index=df.index)

    X = pd.concat([num_df, cat_oh], axis=1).astype(float)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X.values)
    pd.DataFrame(Xs, columns=X.columns).to_csv(ssl_dir/"listing_features.csv", index=False)
    dump(scaler, ssl_dir/"scaler.joblib")
    (ssl_dir/"feature_metadata.json").write_text(json.dumps({
        "num_cols": num_cols, "cat_cols": cat_cols, "cat_topk": meta_top,
        "feature_columns": X.columns.tolist()
    }, ensure_ascii=False, indent=2))

    host_ids = df["host_id"].values

    # 3) 대조학습 데이터로더
    ds = ContrastiveListingDataset(Xs, host_ids, feature_noise=noise, mask_prob=mask)
    dl = DataLoader(ds, batch_size=batch, shuffle=True, num_workers=workers,
                    drop_last=True, pin_memory=(device.type=="cuda"))

    # 4) SSL 학습
    input_dim = Xs.shape[1]
    encoder = MLPEncoder(input_dim, hidden, emb).to(device)
    projector = ProjectionHead(emb, proj_hidden, proj_dim).to(device)
    ckpt_dir = ssl_dir / "checkpoints"
    train_contrastive(encoder, projector, dl, device,
                      epochs=epochs, lr=lr, temperature=temperature,
                      grad_accum_steps=grad_accum, save_every=save_every, out_dir=ckpt_dir)

    # 5) 임베딩 → 호스트 평균 임베딩
    encoder.eval(); E=[]
    with torch.no_grad():
        bs = 4096
        for i in range(0, Xs.shape[0], bs):
            xb = torch.from_numpy(Xs[i:i+bs]).to(device).float()
            E.append(encoder(xb).cpu().numpy())
    E = np.vstack(E)
    emb_df = pd.DataFrame(E); emb_df["host_id"] = host_ids
    host_vec = emb_df.groupby("host_id").mean().sort_index()
    host_vec.to_csv(ssl_dir/"host_vectors_ssl.csv")

    # 6) KPI 집계
    host_kpis = build_host_kpis(df)

    # 7) 약한 라벨(슈퍼호스트) 추출(있으면)
    weak_labels = None
    for cand in ["host_is_superhost","is_superhost"]:
        if cand in df.columns:
            weak_labels = df.groupby("host_id")[cand].first()
            break

    # 8) 점수/구성요소 산출
    score, kpi_score, maybe_ssl_score, Z_kpi, Z_contrib, meta = build_score_with_components(
        host_kpis=host_kpis,
        ssl_host_vec=host_vec,
        weak_labels=weak_labels,
        alpha_ssl=alpha_ssl
    )

    # 9) 등급 + 커트라인
    tiers, cutoffs = cut_into_tiers_with_cutoffs(score, p_master=p_master, p_super=p_super, p_rising=p_rising)

    # 10) 결과 저장(최종 테이블: KPI/SSL/최종점수/등급)
    result = pd.DataFrame({
        "host_id": score.index,
        "score": score.values,
        "kpi_score": kpi_score.reindex(score.index).values,
        "tier": tiers.reindex(score.index).values
    }).set_index("host_id")

    if maybe_ssl_score is not None:
        result["ssl_score"] = maybe_ssl_score.reindex(score.index)
    # KPI 원본도 붙여서 한번에 보기
    result = result.join(host_kpis, how="left")

    result.reset_index().to_csv(tier_dir/"host_tiers_ranked.csv", index=False)

    # 11) KPI 기여도 저장(각 KPI가 점수에 얼마나 기여했는가)
    contrib = Z_contrib.copy()
    contrib["kpi_score"] = kpi_score
    contrib["final_score"] = score
    contrib["tier"] = tiers
    contrib.reset_index().to_csv(tier_dir/"kpi_contributions.csv", index=False)

    # 12) 등급별 KPI 평균 + 단조성 체크
    report, violations = tier_report(host_kpis, tiers)
    report.to_csv(tier_dir/"tier_kpi_report.csv")

    # 13) 커트라인/분포 메타 저장
    (tier_dir/"score_cutoffs.json").write_text(json.dumps(cutoffs, ensure_ascii=False, indent=2))
    distro = result["tier"].value_counts().to_dict()
    (tier_dir/"tier_distribution.json").write_text(json.dumps(distro, ensure_ascii=False, indent=2))

    # 14) 증거 요약 리포트(evidence_report.md)
    lines = []
    lines.append("# Tiering Evidence Report\n")
    lines.append("## 1) Percentile Cutoffs (최종 점수 기준)\n")
    for k,v in cutoffs.items():
        lines.append(f"- {k}: {v:.6f}")
    lines.append("\n## 2) Tier Distribution (개수)\n")
    for k,v in distro.items():
        lines.append(f"- {k}: {v}")
    lines.append("\n## 3) KPI Weights & Directions\n")
    for k,v in meta["weights"].items():
        sign = "↑(+)" if meta["dir_map"][k] > 0 else "↓(-)"
        lines.append(f"- {k}: weight={v}, dir={sign}")
    lines.append("\n## 4) Monotonicity Check on Tier Averages\n")
    if violations:
        lines.append("⚠ 단조성 위반:")
        for v in violations: lines.append(f"- {v}")
    else:
        lines.append("✅ 주요 KPI 단조성 만족")
    lines.append("\n## 5) Sample Hosts per Tier (Top by Score)\n")
    for t in ["Master","Super","Rising","Rookie"]:
        sub = result[result["tier"]==t].sort_values("score", ascending=False).head(5)
        if len(sub)==0: continue
        lines.append(f"### {t} (상위 5)\n")
        lines.append(sub.reset_index()[["host_id","score","kpi_score"] + ([ "ssl_score"] if "ssl_score" in sub.columns else [])].to_markdown(index=False))
        lines.append("")
    (tier_dir/"evidence_report.md").write_text("\n".join(lines), encoding="utf-8")

    # 15) 콘솔 출력
    print("\n=== 완료 ===")
    print("출력 폴더:", out_dir.resolve())
    print(" [SSL]  - listing_features.csv\n   - scaler.joblib\n   - feature_metadata.json\n   - checkpoints/ssl_epoch*.pt\n   - host_vectors_ssl.csv")
    print(" [Tier] - host_tiers_ranked.csv\n   - kpi_contributions.csv\n   - tier_kpi_report.csv\n   - score_cutoffs.json\n   - tier_distribution.json\n   - evidence_report.md")
    print("\n티어 분포:")
    print(pd.Series(distro))
    if violations:
        print("\n⚠ 단조성 위반:")
        for v in violations: print(" -", v)
    else:
        print("\n✅ 주요 KPI 단조성 만족")


# -------------------- CLI --------------------
if __name__ == "__main__":
    ap = argparse.ArgumentParser()
    ap.add_argument("--raw_csv", type=str, default="./combined_dataset.csv")
    ap.add_argument("--out_dir", type=str, default="./out")
    # SSL
    ap.add_argument("--epochs", type=int, default=50)
    ap.add_argument("--batch", type=int, default=512)
    ap.add_argument("--lr", type=float, default=1e-3)
    ap.add_argument("--temperature", type=float, default=0.5)
    ap.add_argument("--hidden", type=int, default=512)
    ap.add_argument("--emb", type=int, default=128)
    ap.add_argument("--proj_hidden", type=int, default=256)
    ap.add_argument("--proj_dim", type=int, default=64)
    ap.add_argument("--noise", type=float, default=0.05)
    ap.add_argument("--mask", type=float, default=0.05)
    ap.add_argument("--grad_accum", type=int, default=1)
    ap.add_argument("--save_every", type=int, default=10)
    # Ranked cut
    ap.add_argument("--p_master", type=float, default=0.05)
    ap.add_argument("--p_super", type=float, default=0.15)
    ap.add_argument("--p_rising", type=float, default=0.30)
    ap.add_argument("--alpha_ssl", type=float, default=0.5)
    # Exec
    ap.add_argument("--seed", type=int, default=42)
    ap.add_argument("--cpu", action="store_true")
    ap.add_argument("--workers", type=int, default=4)

    args, _ = ap.parse_known_args()  # 주피터 인자 무시
    run_ssl_ranked_with_evidence(**vars(args))
```

```python
# -*- coding: utf-8 -*-
"""
End-to-End: SSL(대조학습) + KPI/SSL 점수화 + 퍼센타일 4등급 + '증거 패키지' 생성
- 주피터/코랩 호환(parse_known_args)
- 함수형 API: run_ssl_ranked_with_evidence(...)

산출물:
  out/
    ssl/
      listing_features.csv, scaler.joblib, feature_metadata.json
      checkpoints/ssl_epoch*.pt
      host_vectors_ssl.csv
    ranked_tiers/
      host_tiers_ranked.csv
      tier_kpi_report.csv
      score_cutoffs.json
      kpi_contributions.csv
      tier_distribution.json
      evidence_report.md
"""

import os, json, random, argparse
from pathlib import Path
from typing import Dict, List, Optional, Tuple

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, Ridge
from joblib import dump

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader


# -------------------- utils --------------------
def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

def parse_price(val):
    if pd.isna(val): return np.nan
    if isinstance(val, (int, float)): return float(val)
    s = str(val)
    for sym in ["$", "€", "₩", ","]:
        s = s.replace(sym, "")
    try: return float(s)
    except: return np.nan

def topk_one_hot(s: pd.Series, k: int, name: str):
    s = s.fillna("UNKNOWN").astype(str)
    top = s.value_counts().nlargest(k).index
    s2 = s.where(s.isin(top), other="OTHER")
    oh = pd.get_dummies(s2, prefix=name)
    return oh, list(top)

def bool_from_any(x):
    return str(x).lower() in {"1","t","true","y","yes"}

def zscore_series(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").astype(float)
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def safe_log1p(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").fillna(0.0)
    return np.log1p(np.clip(s, 0, None))


# -------------------- dataset (contrastive SSL) --------------------
class ContrastiveListingDataset(Dataset):
    """같은 host_id의 두 리스팅(또는 증강뷰)을 positive pair로 구성"""
    def __init__(self, X: np.ndarray, host_ids: np.ndarray,
                 feature_noise=0.05, mask_prob=0.05):
        self.X = X.astype(np.float32)
        self.noise = feature_noise
        self.mask = mask_prob
        self.host_to_rows: Dict[int, List[int]] = {}
        for i, h in enumerate(host_ids):
            self.host_to_rows.setdefault(int(h), []).append(i)
        self.hosts = list(self.host_to_rows.keys())

    def __len__(self): return len(self.hosts)

    def _augment(self, v: np.ndarray) -> np.ndarray:
        u = v.copy()
        if self.noise > 0:
            u = u + self.noise * np.random.randn(*u.shape).astype(np.float32)
        if self.mask > 0:
            m = (np.random.rand(*u.shape) < self.mask)
            u[m] = 0.0
        return u.astype(np.float32)

    def __getitem__(self, idx):
        h = self.hosts[idx]
        rows = self.host_to_rows[h]
        if len(rows) >= 2:
            i1, i2 = np.random.choice(rows, size=2, replace=False)
        else:
            i1 = i2 = rows[0]
        return torch.from_numpy(self._augment(self.X[i1])), torch.from_numpy(self._augment(self.X[i2]))


# -------------------- models --------------------
class MLPEncoder(nn.Module):
    def __init__(self, d_in, d_h=512, d_emb=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_emb)
        )
    def forward(self, x):
        return self.net(x)

class ProjectionHead(nn.Module):
    def __init__(self, d_in=128, d_h=256, d_out=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_out)
        )
    def forward(self, h):
        z = self.net(h)
        return F.normalize(z, dim=1)


# -------------------- training (contrastive InfoNCE) --------------------
def train_contrastive(encoder, projector, dataloader, device,
                      epochs=50, lr=1e-3, temperature=0.5,
                      grad_accum_steps=1, save_every=10, out_dir=Path("./ssl_ckpt")):
    out_dir.mkdir(parents=True, exist_ok=True)
    opt = torch.optim.Adam(list(encoder.parameters()) + list(projector.parameters()), lr=lr)
    for ep in range(1, epochs+1):
        encoder.train(); projector.train()
        running = 0.0
        for step, (x1, x2) in enumerate(dataloader, start=1):
            x1, x2 = x1.to(device), x2.to(device)
            z1, z2 = projector(encoder(x1)), projector(encoder(x2))
            logits = (z1 @ z2.t()) / temperature
            y = torch.arange(logits.size(0), device=device)
            loss = 0.5 * (F.cross_entropy(logits, y) + F.cross_entropy(logits.t(), y))
            (loss/grad_accum_steps).backward()
            if step % grad_accum_steps == 0:
                opt.step(); opt.zero_grad()
            running += loss.item()
        print(f"[Epoch {ep:03d}] loss={running/len(dataloader):.4f}")
        if ep % save_every == 0 or ep == epochs:
            torch.save({
                "encoder": encoder.state_dict(),
                "proj": projector.state_dict(),
                "cfg": {"epochs": epochs, "lr": lr, "temperature": temperature,
                        "grad_accum_steps": grad_accum_steps}
            }, out_dir / f"ssl_epoch{ep}.pt")


# -------------------- KPI 집계 --------------------
def build_host_kpis(df: pd.DataFrame) -> pd.DataFrame:
    agg_defs = {
        "avg_rating": ("review_scores_rating", "mean", True),
        "sum_reviews": ("number_of_reviews", "sum", True),
        "avg_reviews_per_month": ("reviews_per_month", "mean", True),
        "median_price": ("price", "median", False),
        "availability_30": ("availability_30", "mean", False),
        "availability_90": ("availability_90", "mean", False),
        "availability_365": ("availability_365", "mean", False),
    }
    if "price" in df.columns:
        df = df.copy()
        df["price"] = df["price"].apply(parse_price)

    out = pd.DataFrame({"host_id": df["host_id"]}).drop_duplicates().set_index("host_id")
    for name, (col, how, _) in agg_defs.items():
        if col not in df.columns: continue
        if how == "mean":
            s = df.groupby("host_id")[col].mean()
        elif how == "sum":
            s = df.groupby("host_id")[col].sum()
        elif how == "median":
            s = df.groupby("host_id")[col].median()
        else:
            continue
        out[name] = s
    return out


# -------------------- 점수 산출 (KPI + SSL 약지도 블렌딩) --------------------
def build_score_with_components(host_kpis: pd.DataFrame,
                                ssl_host_vec: Optional[pd.DataFrame],
                                weak_labels: Optional[pd.Series],
                                weights: Optional[dict]=None,
                                alpha_ssl=0.5):
    """
    반환:
      score: 최종 점수(Series)
      kpi_score: KPI 점수(Series)
      ssl_score: SSL-약지도 점수(Series 또는 None)
      Z: KPI z-score 테이블(방향 보정 전)
      Z_oriented: 방향/가중 반영된 KPI 기여도 테이블
      meta: dir_map, weights 등 메타
    """
    dir_map = {
        "avg_rating": +1, "sum_reviews": +1, "avg_reviews_per_month": +1,
        "median_price": -1, "availability_30": -1, "availability_90": -1, "availability_365": -1,
    }
    use_cols = [c for c in host_kpis.columns if c in dir_map]
    if not use_cols:
        raise ValueError("KPI가 없습니다. 평점/리뷰/가용일수 등 열이 데이터에 있는지 확인하세요.")

    # KPI z-score
    Z = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        s = host_kpis[c]
        if c in {"sum_reviews"}:
            s = safe_log1p(s)
        z = zscore_series(s.fillna(s.median()))
        Z[c] = z

    # 가중치 (평점 2배 기본)
    if weights is None:
        weights = {c: 1.0 for c in use_cols}
        if "avg_rating" in weights: weights["avg_rating"] = 2.0

    # 방향 보정 + 기여도
    Z_oriented = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        Z_oriented[f"contrib_{c}"] = dir_map[c] * weights[c] * Z[c]

    kpi_score = Z_oriented.sum(axis=1)
    kpi_score = zscore_series(kpi_score)

    # SSL-약지도 점수
    ssl_score = None
    if ssl_host_vec is not None and weak_labels is not None and weak_labels.notna().any():
        idx = host_kpis.index.intersection(ssl_host_vec.index).intersection(weak_labels.index)
        if len(idx) > 5:
            X_ssl = ssl_host_vec.loc[idx].select_dtypes(include=[np.number]).values
            y = weak_labels.loc[idx].apply(bool_from_any).astype(int).values
            scaler = StandardScaler(); Xs = scaler.fit_transform(X_ssl)
            try:
                clf = LogisticRegression(max_iter=1000); clf.fit(Xs, y)
                s = clf.predict_proba(Xs)[:,1]
            except Exception:
                reg = Ridge(alpha=1.0); reg.fit(Xs, y)
                s = reg.predict(Xs)
            s = (s - s.mean()) / (s.std() + 1e-9)
            ssl_score = pd.Series(s, index=idx, name="ssl_score")

    # 최종 점수 블렌딩
    final = kpi_score.copy()
    if ssl_score is not None:
        final.loc[ssl_score.index] = alpha_ssl*kpi_score.loc[ssl_score.index] + (1-alpha_ssl)*ssl_score

    meta = {"dir_map": dir_map, "weights": weights}
    return final.rename("score"), kpi_score.rename("kpi_score"), ssl_score, Z[use_cols], Z_oriented, meta


# -------------------- 퍼센타일 등급 컷 --------------------
def cut_into_tiers_with_cutoffs(score: pd.Series, p_master=0.05, p_super=0.15, p_rising=0.30):
    s = score.sort_values(ascending=False)
    n = len(s)
    n_master = max(1, int(round(p_master*n)))
    n_super  = max(1, int(round(p_super*n)))
    n_rising = max(1, int(round(p_rising*n)))
    idx_master = s.index[:n_master]
    idx_super  = s.index[n_master:n_master+n_super]
    idx_rising = s.index[n_master+n_super:n_master+n_super+n_rising]
    idx_rookie = s.index[n_master+n_super+n_rising:]
    tier = pd.Series(index=s.index, dtype=object)
    tier.loc[idx_master] = "Master"
    tier.loc[idx_super]  = "Super"
    tier.loc[idx_rising] = "Rising"
    tier.loc[idx_rookie] = "Rookie"

    # 커트라인(실제 수치) 계산
    cut_master = s.iloc[n_master-1] if n_master>0 else s.max()
    cut_super  = s.iloc[n_master+n_super-1] if n_super>0 else s.iloc[n_master-1]
    cut_rising = s.iloc[n_master+n_super+n_rising-1] if n_rising>0 else s.iloc[n_master+n_super-1]
    cutoffs = {
        "Master_min_score": float(cut_master),
        "Super_min_score":  float(cut_super),
        "Rising_min_score": float(cut_rising),
    }
    return tier, cutoffs


# -------------------- 등급 리포트 & 단조성 체크 --------------------
def tier_report(df_host, tiers,
                kpi_cols=("avg_rating","sum_reviews","avg_reviews_per_month","median_price","availability_30","availability_90","availability_365")):
    rep = df_host.copy(); rep["tier"] = tiers
    g = rep.groupby("tier")[list(c for c in kpi_cols if c in rep.columns)].mean()
    order = ["Rookie","Rising","Super","Master"]; g = g.reindex(order)
    dir_map = {"avg_rating":"+","sum_reviews":"+","avg_reviews_per_month":"+",
               "median_price":"-","availability_30":"-","availability_90":"-","availability_365":"-"}
    violations = []
    for c in g.columns:
        d = dir_map.get(c, "+"); vals = g[c].values
        if np.isnan(vals).any(): continue
        if d == "+" and not np.all(np.diff(vals) >= -1e-9):
            violations.append(f"{c}: 기대 ↑ 단조 위반")
        if d == "-" and not np.all(np.diff(vals) <=  1e-9):
            violations.append(f"{c}: 기대 ↓ 단조 위반")
    return g, violations


# -------------------- 보기 좋은 KPI 평균 표 --------------------
def pretty_tier_kpi_table(host_kpis: pd.DataFrame, tiers: pd.Series) -> pd.DataFrame:
    order = ["Rookie","Rising","Super","Master"]
    kpi_cols = ["avg_rating","sum_reviews","avg_reviews_per_month",
                "median_price","availability_30","availability_90","availability_365"]
    kpi_cols = [c for c in kpi_cols if c in host_kpis.columns]
    rep = host_kpis.copy()
    rep["tier"] = tiers
    tbl = rep.groupby("tier")[kpi_cols].mean().reindex(order)
    # 보기 좋은 출력
    pd.set_option("display.max_columns", None)
    pd.set_option("display.width", 120)
    print("\n등급별 KPI 평균:")
    print(tbl)
    return tbl


# -------------------- End-to-End + Evidence --------------------
def run_ssl_ranked_with_evidence(
    raw_csv="combined_dataset.csv",
    out_dir="out",
    # SSL 학습
    epochs=50, batch=512, lr=1e-3, temperature=0.5,
    hidden=512, emb=128, proj_hidden=256, proj_dim=64,
    noise=0.05, mask=0.05, grad_accum=1, save_every=10,
    # Ranked tier cut
    p_master=0.05, p_super=0.15, p_rising=0.30, alpha_ssl=0.5,
    # Exec
    seed=42, cpu=False, workers=4
):
    set_seed(seed)
    device = torch.device("cuda" if torch.cuda.is_available() and not cpu else "cpu")
    print("Device:", device)

    raw_csv = Path(raw_csv)
    out_dir = Path(out_dir); out_dir.mkdir(parents=True, exist_ok=True)
    ssl_dir = out_dir / "ssl"
    tier_dir = out_dir / "ranked_tiers"
    ssl_dir.mkdir(exist_ok=True, parents=True)
    tier_dir.mkdir(exist_ok=True, parents=True)

    # 1) 데이터 로드/정리
    df = pd.read_csv(raw_csv)
    print("Data shape:", df.shape)
    if "price" in df.columns: df["price"] = df["price"].apply(parse_price)
    assert "host_id" in df.columns, "host_id 컬럼이 필요합니다."

    # 2) 리스팅 피처 엔지니어링
    num_cols_cand = [
        "price","number_of_reviews","reviews_per_month","calculated_host_listings_count",
        "minimum_nights","maximum_nights","availability_30","availability_60",
        "availability_90","availability_365","review_scores_rating",
        "review_scores_accuracy","review_scores_cleanliness","review_scores_checkin",
        "review_scores_communication","review_scores_location","review_scores_value"
    ]
    num_cols = [c for c in num_cols_cand if c in df.columns]
    cat_cols_cand = ["room_type","property_type","neighbourhood_cleansed",
                     "bathrooms_text","instant_bookable","host_response_time"]
    cat_cols = [c for c in cat_cols_cand if c in df.columns]

    num_df = df[num_cols].copy().apply(pd.to_numeric, errors="coerce")
    num_df = num_df.fillna(num_df.median())

    oh_parts, meta_top = [], {}
    for c in cat_cols:
        k = 30 if c == "property_type" else (50 if c == "neighbourhood_cleansed" else 10)
        oh, top_vals = topk_one_hot(df[c], k=k, name=c)
        oh_parts.append(oh); meta_top[c] = {"k": k, "top_values": top_vals}
    cat_oh = pd.concat(oh_parts, axis=1) if oh_parts else pd.DataFrame(index=df.index)

    X = pd.concat([num_df, cat_oh], axis=1).astype(float)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X.values)
    pd.DataFrame(Xs, columns=X.columns).to_csv(ssl_dir/"listing_features.csv", index=False)
    dump(scaler, ssl_dir/"scaler.joblib")
    (ssl_dir/"feature_metadata.json").write_text(json.dumps({
        "num_cols": num_cols, "cat_cols": cat_cols, "cat_topk": meta_top,
        "feature_columns": X.columns.tolist()
    }, ensure_ascii=False, indent=2))

    host_ids = df["host_id"].values

    # 3) 대조학습 데이터로더
    ds = ContrastiveListingDataset(Xs, host_ids, feature_noise=noise, mask_prob=mask)
    dl = DataLoader(ds, batch_size=batch, shuffle=True, num_workers=workers,
                    drop_last=True, pin_memory=(device.type=="cuda"))

    # 4) SSL 학습
    input_dim = Xs.shape[1]
    encoder = MLPEncoder(input_dim, hidden, emb).to(device)
    projector = ProjectionHead(emb, proj_hidden, proj_dim).to(device)
    ckpt_dir = ssl_dir / "checkpoints"
    train_contrastive(encoder, projector, dl, device,
                      epochs=epochs, lr=lr, temperature=temperature,
                      grad_accum_steps=grad_accum, save_every=save_every, out_dir=ckpt_dir)

    # 5) 임베딩 → 호스트 평균 임베딩
    encoder.eval(); E=[]
    with torch.no_grad():
        bs = 4096
        for i in range(0, Xs.shape[0], bs):
            xb = torch.from_numpy(Xs[i:i+bs]).to(device).float()
            E.append(encoder(xb).cpu().numpy())
    E = np.vstack(E)
    emb_df = pd.DataFrame(E); emb_df["host_id"] = host_ids
    host_vec = emb_df.groupby("host_id").mean().sort_index()
    host_vec.to_csv(ssl_dir/"host_vectors_ssl.csv")

    # 6) KPI 집계
    host_kpis = build_host_kpis(df)

    # 7) 약한 라벨(슈퍼호스트) 추출(있으면)
    weak_labels = None
    for cand in ["host_is_superhost","is_superhost"]:
        if cand in df.columns:
            weak_labels = df.groupby("host_id")[cand].first()
            break

    # 8) 점수/구성요소 산출
    score, kpi_score, maybe_ssl_score, Z_kpi, Z_contrib, meta = build_score_with_components(
        host_kpis=host_kpis,
        ssl_host_vec=host_vec,
        weak_labels=weak_labels,
        alpha_ssl=alpha_ssl
    )

    # 9) 등급 + 커트라인
    tiers, cutoffs = cut_into_tiers_with_cutoffs(score, p_master=p_master, p_super=p_super, p_rising=p_rising)

    # 10) 결과 저장(최종 테이블: KPI/SSL/최종점수/등급)
    result = pd.DataFrame({
        "host_id": score.index,
        "score": score.values,
        "kpi_score": kpi_score.reindex(score.index).values,
        "tier": tiers.reindex(score.index).values
    }).set_index("host_id")
    if maybe_ssl_score is not None:
        result["ssl_score"] = maybe_ssl_score.reindex(score.index)
    result = result.join(host_kpis, how="left")
    result.reset_index().to_csv(tier_dir/"host_tiers_ranked.csv", index=False)

    # 11) KPI 기여도 저장
    contrib = Z_contrib.copy()
    contrib["kpi_score"] = kpi_score
    contrib["final_score"] = score
    contrib["tier"] = tiers
    contrib.reset_index().to_csv(tier_dir/"kpi_contributions.csv", index=False)

    # 12) 등급별 KPI 평균 + 단조성 체크 + 콘솔 예쁘게 출력
    report, violations = tier_report(host_kpis, tiers)
    report.to_csv(tier_dir/"tier_kpi_report.csv")
    # pretty print (콘솔에 보기 좋게)
    _ = pretty_tier_kpi_table(host_kpis, tiers)

    # 13) 커트라인/분포 메타 저장
    (tier_dir/"score_cutoffs.json").write_text(json.dumps(cutoffs, ensure_ascii=False, indent=2))
    distro = result["tier"].value_counts().to_dict()
    (tier_dir/"tier_distribution.json").write_text(json.dumps(distro, ensure_ascii=False, indent=2))

    # 14) 증거 요약 리포트
    lines = []
    lines.append("# Tiering Evidence Report\n")
    lines.append("## 1) Percentile Cutoffs (최종 점수 기준)\n")
    for k,v in cutoffs.items():
        lines.append(f"- {k}: {v:.6f}")
    lines.append("\n## 2) Tier Distribution (개수)\n")
    for k,v in distro.items():
        lines.append(f"- {k}: {v}")
    lines.append("\n## 3) KPI Weights & Directions\n")
    for k,v in meta["weights"].items():
        sign = "↑(+)" if meta["dir_map"][k] > 0 else "↓(-)"
        lines.append(f"- {k}: weight={v}, dir={sign}")
    lines.append("\n## 4) Monotonicity Check on Tier Averages\n")
    if violations:
        lines.append("⚠ 단조성 위반:")
        for v in violations: lines.append(f"- {v}")
    else:
        lines.append("✅ 주요 KPI 단조성 만족")
    lines.append("\n## 5) Sample Hosts per Tier (Top by Score)\n")
    for t in ["Master","Super","Rising","Rookie"]:
        sub = result[result["tier"]==t].sort_values("score", ascending=False).head(5)
        if len(sub)==0: continue
        lines.append(f"### {t} (상위 5)\n")
        cols = ["host_id","score","kpi_score"]
        if "ssl_score" in sub.columns: cols.append("ssl_score")
        lines.append(sub.reset_index()[cols].to_markdown(index=False))
        lines.append("")
    (tier_dir/"evidence_report.md").write_text("\n".join(lines), encoding="utf-8")

    # 15) 콘솔 출력
    print("\n=== 완료 ===")
    print("출력 폴더:", out_dir.resolve())
    print(" [SSL]  - listing_features.csv\n   - scaler.joblib\n   - feature_metadata.json\n   - checkpoints/ssl_epoch*.pt\n   - host_vectors_ssl.csv")
    print(" [Tier] - host_tiers_ranked.csv\n   - kpi_contributions.csv\n   - tier_kpi_report.csv\n   - score_cutoffs.json\n   - tier_distribution.json\n   - evidence_report.md")
    print("\n티어 분포:")
    print(pd.Series(distro))
    if violations:
        print("\n⚠ 단조성 위반:")
        for v in violations: print(" -", v)
    else:
        print("\n✅ 주요 KPI 단조성 만족")


# -------------------- CLI --------------------
if __name__ == "__main__":
    ap = argparse.ArgumentParser()
    ap.add_argument("--raw_csv", type=str, default="./combined_dataset.csv")
    ap.add_argument("--out_dir", type=str, default="./out")
    # SSL
    ap.add_argument("--epochs", type=int, default=50)
    ap.add_argument("--batch", type=int, default=512)
    ap.add_argument("--lr", type=float, default=1e-3)
    ap.add_argument("--temperature", type=float, default=0.5)
    ap.add_argument("--hidden", type=int, default=512)
    ap.add_argument("--emb", type=int, default=128)
    ap.add_argument("--proj_hidden", type=int, default=256)
    ap.add_argument("--proj_dim", type=int, default=64)
    ap.add_argument("--noise", type=float, default=0.05)
    ap.add_argument("--mask", type=float, default=0.05)
    ap.add_argument("--grad_accum", type=int, default=1)
    ap.add_argument("--save_every", type=int, default=10)
    # Ranked cut
    ap.add_argument("--p_master", type=float, default=0.05)
    ap.add_argument("--p_super", type=float, default=0.15)
    ap.add_argument("--p_rising", type=float, default=0.30)
    ap.add_argument("--alpha_ssl", type=float, default=0.5)
    # Exec
    ap.add_argument("--seed", type=int, default=42)
    ap.add_argument("--cpu", action="store_true")
    ap.add_argument("--workers", type=int, default=4)

    args, _ = ap.parse_known_args()  # 주피터 인자 무시
    run_ssl_ranked_with_evidence(**vars(args))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.covariance import EmpiricalCovariance
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# ============================================
#  1. 데이터 로드
# ============================================
df = pd.read_csv("/content/combined_dataset.csv")

# host_is_superhost -> binary
df = df[df['host_is_superhost'].notna()]
df['host_is_superhost'] = df['host_is_superhost'].map({'t': 1, 'f': 0})

# feature 선택 + 전처리
features = ['review_scores_rating', 'number_of_reviews',
            'reviews_per_month', 'host_response_rate', 'host_acceptance_rate']
for col in ['host_response_rate','host_acceptance_rate']:
    df[col] = df[col].str.rstrip('%').astype(float)

df = df[features + ['host_is_superhost']].dropna()
X = df[features].values
y = df['host_is_superhost'].values

# 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================
#  2. Mahalanobis Distance 기반 score 계산
# ============================================
cov = EmpiricalCovariance().fit(X_scaled)
mean_true = X_scaled[y==1].mean(axis=0)
mean_false = X_scaled[y==0].mean(axis=0)

def mahalanobis_distance(x, mean):
    diff = x - mean
    return np.sqrt(np.dot(np.dot(diff, cov.precision_), diff.T))

dist_true = np.apply_along_axis(mahalanobis_distance, 1, X_scaled, mean_true)
dist_false = np.apply_along_axis(mahalanobis_distance, 1, X_scaled, mean_false)

score = 1 - (dist_true / (dist_true + dist_false))
df['superhost_score'] = score

# ============================================
#  3. Q1, Q2, Q3 기준으로 Rookie / Rising / UpperGroup 분리
# ============================================
q1, q2, q3 = np.quantile(score, [0.25, 0.5, 0.75])
df['tier_tmp'] = np.where(score <= q1, "Rookie",
                   np.where(score <= q2, "Rising", "UpperGroup"))

# ============================================
#  4. UpperGroup만 KMeans로 Super/Master 세분화
# ============================================
upper_df = df[df['tier_tmp']=="UpperGroup"].copy()
upper_scores = upper_df['superhost_score'].values.reshape(-1,1)

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
upper_clusters = kmeans.fit_predict(upper_scores)
centers = kmeans.cluster_centers_.flatten()

# 클러스터 중심 정렬 (작은쪽 = Super, 큰쪽 = Master)
low_cluster = np.argmin(centers)
upper_df['tier'] = np.where(upper_clusters==low_cluster, "Super", "Master")

# ============================================
#  5. 최종 tier 병합
# ============================================
df = df.merge(upper_df[['superhost_score','tier']],
              on='superhost_score', how='left')
df['tier'] = df.apply(lambda x: x['tier']
                      if pd.notna(x['tier'])
                      else x['tier_tmp'], axis=1)

# ============================================
#  6. 결과 확인
# ============================================
print("=== Q1/Q2/Q3 경계값 ===")
print(f"Q1: {q1:.4f}, Q2: {q2:.4f}, Q3: {q3:.4f}\n")

print("=== 등급별 분포 ===")
print(df['tier'].value_counts(normalize=True)*100)

print("\n=== 등급별 평균 KPI ===")
print(df.groupby('tier')[['superhost_score','review_scores_rating','number_of_reviews']].mean())

# 시각화
plt.figure(figsize=(10,5))
plt.hist(score, bins=50, color='skyblue', edgecolor='black')
plt.axvline(q1, color='purple', linestyle='--', label='Q1')
plt.axvline(q2, color='orange', linestyle='--', label='Q2')
plt.axvline(q3, color='green', linestyle='--', label='Q3')
plt.title("Superhost Score Distribution with Quartiles")
plt.xlabel("Superhost Score")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# ============================================
# Colab-ready: 중요도 기반 피처선택 -> KMeans(k=4) -> 등급라벨(루키/라이징/슈퍼/마스터)
# - review_scores_rating 예측용 RF로 feature importance 산출
# - 누적 90%까지 상위 피처 선택(최소 3, 최대 8)
# - 선택 피처 표준화 후 KMeans(k=4) 군집
# - 각 군집의 평균 rating으로 순서를 매겨 등급 라벨 부여
# - 결과/요약/중요도 CSV 저장
# ============================================

import os, numpy as np, pandas as pd
from typing import List

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# 0) 경로 설정 (Colab에 업로드했으면 /content/xxx.csv 로 바꾸세요)
CSV_PATH = "/content/combined_dataset.csv"   # <-- 이 줄만 네 환경에 맞게 수정

# 1) 데이터 로드
df = pd.read_csv(CSV_PATH)

TARGET = "review_scores_rating"
CANDIDATES = [
    "number_of_reviews",
    "reviews_per_month",
    "host_response_rate",
    "host_acceptance_rate",
    "host_is_superhost",
    "host_total_listings_count",
    "minimum_nights",
    "availability_365",
    "instant_bookable",
    "calculated_host_listings_count",
]
CANDIDATES = [c for c in CANDIDATES if c in df.columns]
assert TARGET in df.columns, f"'{TARGET}' 컬럼이 없습니다. 현재 컬럼: {df.columns.tolist()[:30]}"

# 2) 전처리: 퍼센트→숫자, t/f→0/1, 숫자화
for col in ["host_response_rate", "host_acceptance_rate"]:
    if col in df.columns and df[col].dtype == object:
        df[col] = pd.to_numeric(df[col].str.rstrip("%"), errors="coerce")

for col in ["host_is_superhost", "instant_bookable"]:
    if col in df.columns:
        if df[col].dtype == object:
            df[col] = df[col].map({"t": 1, "f": 0, "True": 1, "False": 0})
        df[col] = pd.to_numeric(df[col], errors="coerce")

for col in CANDIDATES + [TARGET]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# 타겟 결측 제거 (군집 라벨 순서를 정하려면 타겟이 필요)
work = df[CANDIDATES + [TARGET]].dropna(subset=[TARGET]).copy()

# ---- (KeyError fix) id/host_id/name를 work에 안전하게 합치기 ----
for col in ["id", "host_id", "name"]:
    if col in df.columns:
        work[col] = df[col]

# 3) RF 중요도
X = work[CANDIDATES]
y = work[TARGET]
pre = ColumnTransformer([
    ("num", SimpleImputer(strategy="median"), CANDIDATES)
], remainder="drop")
X_imp = pre.fit_transform(X)

rf = RandomForestRegressor(
    n_estimators=200,     # 속도-안정 절충 (느리면 100으로 낮추세요)
    random_state=42,
    n_jobs=-1,
    min_samples_leaf=2
)
rf.fit(X_imp, y)

imp = (
    pd.DataFrame({"feature": CANDIDATES, "importance": rf.feature_importances_})
    .sort_values("importance", ascending=False)
    .reset_index(drop=True)
)
imp["cum"] = imp["importance"].cumsum()

# 누적 90%까지 + 최소 3, 최대 8
selected = imp[imp["cum"] <= 0.90]["feature"].tolist()
if len(selected) < 3:
    selected = imp["feature"].head(3).tolist()
if len(selected) > 8:
    selected = imp["feature"].head(8).tolist()

print("선택 피처:", selected)

# 4) KMeans(k=4) 군집
cluster_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("sc", StandardScaler())
])
X_sel = work[selected]
X_clu = cluster_pipe.fit_transform(X_sel)

kmeans = KMeans(n_clusters=4, random_state=42, n_init=20)
clusters = kmeans.fit_predict(X_clu)
work["cluster"] = clusters

# 5) 평균 rating 기준 순서화 → 등급 라벨
stats = (
    work.groupby("cluster")[TARGET]
        .agg(["count","mean","std","median","min","max"])
        .sort_values("mean")           # 낮은 → 루키, 높은 → 마스터
        .reset_index()
)
order = stats["cluster"].tolist()
rank_map = {c: r for r, c in enumerate(order)}
labels = ["루키", "라이징", "슈퍼", "마스터"]

work["cluster_rank"] = work["cluster"].map(rank_map)
work["tier_label"]    = work["cluster_rank"].map(dict(enumerate(labels)))

# 6) 저장물 구성 & 저장
id_cols = [c for c in ["id","host_id","name"] if c in work.columns]  # <- work 기준으로 안전하게
OUT_COLS = id_cols + selected + [TARGET, "cluster", "cluster_rank", "tier_label"]
# 존재하는 컬럼만 최종 확인 (예방 안전장치)
OUT_COLS = [c for c in OUT_COLS if c in work.columns]

out = work[OUT_COLS].copy()

tier_summary = (
    out.groupby("tier_label")[TARGET]
      .agg(n="count", mean="mean", std="std",
           q1=lambda s: s.quantile(0.25),
           median="median",
           q3=lambda s: s.quantile(0.75))
      .reset_index()
      .sort_values("mean")
)
feature_importance = imp[["feature","importance"]].copy()

SAVE_DIR = "/content"  # Colab 기본 폴더
OUT_PATH = os.path.join(SAVE_DIR, "host_tiers_ranked.csv")
SUM_PATH = os.path.join(SAVE_DIR, "tier_summary.csv")
IMP_PATH = os.path.join(SAVE_DIR, "feature_importance.csv")

out.to_csv(OUT_PATH, index=False)
tier_summary.to_csv(SUM_PATH, index=False)
feature_importance.to_csv(IMP_PATH, index=False)

print("\n=== 상위 중요도 피처 (Top 10) ===")
print(feature_importance.head(10).to_string(index=False))
print("\n=== 등급 요약 (review_scores_rating 기준) ===")
print(tier_summary.to_string(index=False))
print("\nSaved:")
print(OUT_PATH)
print(SUM_PATH)
print(IMP_PATH)
```

```python
# -*- coding: utf-8 -*-
"""
End-to-End: SSL(대조학습) + KPI/SSL 점수화 + 퍼센타일 4등급 + '증거 패키지' + 'metrics.json'
- 주피터/코랩 호환(parse_known_args)
- 함수형 API: run_ssl_ranked_with_evidence(...)

산출물:
  out/
    ssl/
      listing_features.csv, scaler.joblib, feature_metadata.json
      checkpoints/ssl_epoch*.pt
      host_vectors_ssl.csv
    ranked_tiers/
      host_tiers_ranked.csv
      tier_kpi_report.csv
      score_cutoffs.json
      kpi_contributions.csv
      tier_distribution.json
      evidence_report.md
      metrics.json
"""

import os, json, random, argparse
from pathlib import Path
from typing import Dict, List, Optional

import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.metrics import roc_auc_score, average_precision_score, silhouette_score
from joblib import dump

from scipy.stats import spearmanr, kendalltau

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader


# -------------------- utils --------------------
def set_seed(seed: int = 42):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

def parse_price(val):
    if pd.isna(val): return np.nan
    if isinstance(val, (int, float)): return float(val)
    s = str(val)
    for sym in ["$", "€", "₩", ","]:
        s = s.replace(sym, "")
    try: return float(s)
    except: return np.nan

def topk_one_hot(s: pd.Series, k: int, name: str):
    s = s.fillna("UNKNOWN").astype(str)
    top = s.value_counts().nlargest(k).index
    s2 = s.where(s.isin(top), other="OTHER")
    oh = pd.get_dummies(s2, prefix=name)
    return oh, list(top)

def bool_from_any(x):
    return str(x).lower() in {"1","t","true","y","yes"}

def zscore_series(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").astype(float)
    return (s - s.mean()) / (s.std(ddof=0) + 1e-9)

def safe_log1p(s: pd.Series) -> pd.Series:
    s = pd.to_numeric(s, errors="coerce").fillna(0.0)
    return np.log1p(np.clip(s, 0, None))


# -------------------- dataset (contrastive SSL) --------------------
class ContrastiveListingDataset(Dataset):
    """같은 host_id의 두 리스팅(또는 증강뷰)을 positive pair로 구성"""
    def __init__(self, X: np.ndarray, host_ids: np.ndarray,
                 feature_noise=0.05, mask_prob=0.05):
        self.X = X.astype(np.float32)
        self.noise = feature_noise
        self.mask = mask_prob
        self.host_to_rows: Dict[int, List[int]] = {}
        for i, h in enumerate(host_ids):
            self.host_to_rows.setdefault(int(h), []).append(i)
        self.hosts = list(self.host_to_rows.keys())

    def __len__(self): return len(self.hosts)

    def _augment(self, v: np.ndarray) -> np.ndarray:
        u = v.copy()
        if self.noise > 0:
            u = u + self.noise * np.random.randn(*u.shape).astype(np.float32)
        if self.mask > 0:
            m = (np.random.rand(*u.shape) < self.mask)
            u[m] = 0.0
        return u.astype(np.float32)

    def __getitem__(self, idx):
        h = self.hosts[idx]
        rows = self.host_to_rows[h]
        if len(rows) >= 2:
            i1, i2 = np.random.choice(rows, size=2, replace=False)
        else:
            i1 = i2 = rows[0]
        return torch.from_numpy(self._augment(self.X[i1])), torch.from_numpy(self._augment(self.X[i2]))


# -------------------- models --------------------
class MLPEncoder(nn.Module):
    def __init__(self, d_in, d_h=512, d_emb=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_emb)
        )
    def forward(self, x):
        return self.net(x)

class ProjectionHead(nn.Module):
    def __init__(self, d_in=128, d_h=256, d_out=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(d_in, d_h),
            nn.ReLU(inplace=True),
            nn.Linear(d_h, d_out)
        )
    def forward(self, h):
        z = self.net(h)
        return F.normalize(z, dim=1)


# -------------------- training (contrastive InfoNCE) --------------------
def train_contrastive(encoder, projector, dataloader, device,
                      epochs=50, lr=1e-3, temperature=0.5,
                      grad_accum_steps=1, save_every=10, out_dir=Path("./ssl_ckpt")):
    out_dir.mkdir(parents=True, exist_ok=True)
    opt = torch.optim.Adam(list(encoder.parameters()) + list(projector.parameters()), lr=lr)
    for ep in range(1, epochs+1):
        encoder.train(); projector.train()
        running = 0.0
        for step, (x1, x2) in enumerate(dataloader, start=1):
            x1, x2 = x1.to(device), x2.to(device)
            z1, z2 = projector(encoder(x1)), projector(encoder(x2))
            logits = (z1 @ z2.t()) / temperature
            y = torch.arange(logits.size(0), device=device)
            loss = 0.5 * (F.cross_entropy(logits, y) + F.cross_entropy(logits.t(), y))
            (loss/grad_accum_steps).backward()
            if step % grad_accum_steps == 0:
                opt.step(); opt.zero_grad()
            running += loss.item()
        print(f"[Epoch {ep:03d}] loss={running/len(dataloader):.4f}")
        if ep % save_every == 0 or ep == epochs:
            torch.save({
                "encoder": encoder.state_dict(),
                "proj": projector.state_dict(),
                "cfg": {"epochs": epochs, "lr": lr, "temperature": temperature,
                        "grad_accum_steps": grad_accum_steps}
            }, out_dir / f"ssl_epoch{ep}.pt")


# -------------------- KPI 집계 --------------------
def build_host_kpis(df: pd.DataFrame) -> pd.DataFrame:
    agg_defs = {
        "avg_rating": ("review_scores_rating", "mean", True),
        "sum_reviews": ("number_of_reviews", "sum", True),
        "avg_reviews_per_month": ("reviews_per_month", "mean", True),
        "median_price": ("price", "median", False),
        "availability_30": ("availability_30", "mean", False),
        "availability_90": ("availability_90", "mean", False),
        "availability_365": ("availability_365", "mean", False),
    }
    if "price" in df.columns:
        df = df.copy()
        df["price"] = df["price"].apply(parse_price)

    out = pd.DataFrame({"host_id": df["host_id"]}).drop_duplicates().set_index("host_id")
    for name, (col, how, _) in agg_defs.items():
        if col not in df.columns: continue
        if how == "mean":
            s = df.groupby("host_id")[col].mean()
        elif how == "sum":
            s = df.groupby("host_id")[col].sum()
        elif how == "median":
            s = df.groupby("host_id")[col].median()
        else:
            continue
        out[name] = s
    return out


# -------------------- 점수 산출 (KPI + SSL 약지도 블렌딩) --------------------
def build_score_with_components(host_kpis: pd.DataFrame,
                                ssl_host_vec: Optional[pd.DataFrame],
                                weak_labels: Optional[pd.Series],
                                weights: Optional[dict]=None,
                                alpha_ssl=0.5):
    """
    반환:
      score: 최종 점수(Series)
      kpi_score: KPI 점수(Series)
      ssl_score: SSL-약지도 점수(Series 또는 None)
      Z: KPI z-score 테이블(방향 보정 전)
      Z_oriented: 방향/가중 반영된 KPI 기여도 테이블
      meta: dir_map, weights 등 메타
    """
    dir_map = {
        "avg_rating": +1, "sum_reviews": +1, "avg_reviews_per_month": +1,
        "median_price": -1, "availability_30": -1, "availability_90": -1, "availability_365": -1,
    }
    use_cols = [c for c in host_kpis.columns if c in dir_map]
    if not use_cols:
        raise ValueError("KPI가 없습니다. 평점/리뷰/가용일수 등 열이 데이터에 있는지 확인하세요.")

    # KPI z-score
    Z = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        s = host_kpis[c]
        if c in {"sum_reviews"}:
            s = safe_log1p(s)
        z = zscore_series(s.fillna(s.median()))
        Z[c] = z

    # 가중치 (평점 2배 기본)
    if weights is None:
        weights = {c: 1.0 for c in use_cols}
        if "avg_rating" in weights: weights["avg_rating"] = 2.0

    # 방향 보정 + 기여도
    Z_oriented = pd.DataFrame(index=host_kpis.index)
    for c in use_cols:
        Z_oriented[f"contrib_{c}"] = dir_map[c] * weights[c] * Z[c]

    kpi_score = Z_oriented.sum(axis=1)
    kpi_score = zscore_series(kpi_score)

    # SSL-약지도 점수
    ssl_score = None
    if ssl_host_vec is not None and weak_labels is not None and weak_labels.notna().any():
        idx = host_kpis.index.intersection(ssl_host_vec.index).intersection(weak_labels.index)
        if len(idx) > 5:
            X_ssl = ssl_host_vec.loc[idx].select_dtypes(include=[np.number]).values
            y = weak_labels.loc[idx].apply(bool_from_any).astype(int).values
            scaler = StandardScaler(); Xs = scaler.fit_transform(X_ssl)
            try:
                clf = LogisticRegression(max_iter=1000); clf.fit(Xs, y)
                s = clf.predict_proba(Xs)[:,1]
            except Exception:
                reg = Ridge(alpha=1.0); reg.fit(Xs, y)
                s = reg.predict(Xs)
            s = (s - s.mean()) / (s.std() + 1e-9)
            ssl_score = pd.Series(s, index=idx, name="ssl_score")

    # 최종 점수 블렌딩
    final = kpi_score.copy()
    if ssl_score is not None:
        final.loc[ssl_score.index] = alpha_ssl*kpi_score.loc[ssl_score.index] + (1-alpha_ssl)*ssl_score

    meta = {"dir_map": dir_map, "weights": weights}
    return final.rename("score"), kpi_score.rename("kpi_score"), ssl_score, Z[use_cols], Z_oriented, meta


# -------------------- 퍼센타일 등급 컷 --------------------
def cut_into_tiers_with_cutoffs(score: pd.Series, p_master=0.05, p_super=0.15, p_rising=0.30):
    s = score.sort_values(ascending=False)
    n = len(s)
    n_master = max(1, int(round(p_master*n)))
    n_super  = max(1, int(round(p_super*n)))
    n_rising = max(1, int(round(p_rising*n)))
    idx_master = s.index[:n_master]
    idx_super  = s.index[n_master:n_master+n_super]
    idx_rising = s.index[n_master+n_super:n_master+n_super+n_rising]
    idx_rookie = s.index[n_master+n_super+n_rising:]
    tier = pd.Series(index=s.index, dtype=object)
    tier.loc[idx_master] = "Master"
    tier.loc[idx_super]  = "Super"
    tier.loc[idx_rising] = "Rising"
    tier.loc[idx_rookie] = "Rookie"

    # 커트라인(실제 수치) 계산
    cut_master = s.iloc[n_master-1] if n_master>0 else s.max()
    cut_super  = s.iloc[n_master+n_super-1] if n_super>0 else s.iloc[n_master-1]
    cut_rising = s.iloc[n_master+n_super+n_rising-1] if n_rising>0 else s.iloc[n_master+n_super-1]
    cutoffs = {
        "Master_min_score": float(cut_master),
        "Super_min_score":  float(cut_super),
        "Rising_min_score": float(cut_rising),
    }
    return tier, cutoffs


# -------------------- 등급 리포트 & 단조성 체크 --------------------
def tier_report(df_host, tiers,
                kpi_cols=("avg_rating","sum_reviews","avg_reviews_per_month","median_price","availability_30","availability_90","availability_365")):
    rep = df_host.copy(); rep["tier"] = tiers
    g = rep.groupby("tier")[list(c for c in kpi_cols if c in rep.columns)].mean()
    order = ["Rookie","Rising","Super","Master"]; g = g.reindex(order)
    dir_map = {"avg_rating":"+","sum_reviews":"+","avg_reviews_per_month":"+",
               "median_price":"-","availability_30":"-","availability_90":"-","availability_365":"-"}
    violations = []
    for c in g.columns:
        d = dir_map.get(c, "+"); vals = g[c].values
        if np.isnan(vals).any(): continue
        if d == "+" and not np.all(np.diff(vals) >= -1e-9):
            violations.append(f"{c}: 기대 ↑ 단조 위반")
        if d == "-" and not np.all(np.diff(vals) <=  1e-9):
            violations.append(f"{c}: 기대 ↓ 단조 위반")
    return g, violations


# -------------------- 보기 좋은 KPI 평균 표 --------------------
def pretty_tier_kpi_table(host_kpis: pd.DataFrame, tiers: pd.Series) -> pd.DataFrame:
    order = ["Rookie","Rising","Super","Master"]
    kpi_cols = ["avg_rating","sum_reviews","avg_reviews_per_month",
                "median_price","availability_30","availability_90","availability_365"]
    kpi_cols = [c for c in kpi_cols if c in host_kpis.columns]
    rep = host_kpis.copy()
    rep["tier"] = tiers
    tbl = rep.groupby("tier")[kpi_cols].mean().reindex(order)
    # 보기 좋은 출력
    pd.set_option("display.max_columns", None)
    pd.set_option("display.width", 120)
    print("\n등급별 KPI 평균:")
    print(tbl)
    return tbl


# -------------------- 지표 계산 유틸 --------------------
def topk_precision(score: pd.Series, labels: pd.Series, ks=(0.05, 0.10, 0.20)):
    """점수 내림차순 상위 k% 안에서의 양성(슈퍼호스트) 비율"""
    out = {}
    s = score.sort_values(ascending=False)
    y = labels.reindex(s.index).apply(bool_from_any).astype(int)
    n = len(s)
    for k in ks:
        m = max(1, int(round(k*n)))
        top_idx = s.index[:m]
        prec = float(y.loc[top_idx].mean()) if m > 0 else float('nan')
        out[f"precision_at_top_{int(k*100)}pct"] = prec
    return out

def uniformity_metric(X: np.ndarray, t=2.0, num_samples=20000, seed=42):
    """Wang & Isola(2020) 정의를 근사: E[exp(-t * ||xi-xj||^2)]의 로그 (작을수록 좋음)"""
    rng = np.random.default_rng(seed)
    n = X.shape[0]
    if n < 2: return float('nan')
    idx_i = rng.integers(0, n, size=min(num_samples, n))
    idx_j = rng.integers(0, n, size=min(num_samples, n))
    d2 = np.sum((X[idx_i] - X[idx_j])**2, axis=1)
    return float(np.log(np.mean(np.exp(-t * d2)) + 1e-12))

def alignment_metric(E: np.ndarray, host_ids: np.ndarray, num_hosts=200, seed=42):
    """같은 host 내 리스팅 임베딩 쌍들의 평균 L2거리 (작을수록 좋음, 근사)"""
    rng = np.random.default_rng(seed)
    from collections import defaultdict
    h2rows = defaultdict(list)
    for i, h in enumerate(host_ids):
        h2rows[int(h)].append(i)
    hosts = list(h2rows.keys())
    rng.shuffle(hosts)
    hosts = hosts[:min(num_hosts, len(hosts))]
    dists = []
    for h in hosts:
        rows = h2rows[h]
        if len(rows) < 2: continue
        i = rng.choice(rows, size=2, replace=False)
        v1, v2 = E[i[0]], E[i[1]]
        dists.append(np.linalg.norm(v1 - v2))
    if len(dists) == 0: return float('nan')
    return float(np.mean(dists))

def label_rate(labels: Optional[pd.Series]):
    if labels is None or labels.isna().all(): return float('nan')
    y = labels.apply(bool_from_any).astype(int)
    return float(y.mean())


# -------------------- 콘솔 pretty print (metrics) --------------------
def pretty_print_metrics(metrics: dict):
    def p(k, default="(n/a)"):
        v = metrics.get(k, default)
        print(f"{k}: {v}")

    print("\n=== METRICS SUMMARY ===")
    print("\n[SSL 임베딩 품질]")
    p("ssl_uniformity_log")
    p("ssl_alignment_mean_l2")

    print("\n[약라벨 기반 성능(있을 때만)]")
    p("weak_label_rate")
    for k in ["precision_at_top_5pct","precision_at_top_10pct","precision_at_top_20pct"]:
        if k in metrics: p(k)
    for k in ["linear_eval_auroc","linear_eval_auprc"]:
        if k in metrics: p(k)

    print("\n[티어 구조 품질]")
    p("monotonicity_violations")
    if "monotonicity_violation_list" in metrics:
        print("  - violations:", metrics["monotonicity_violation_list"])
    p("silhouette_host_vec_vs_tier")

    print("\n[랭킹 일관성: Score vs KPI 순위상관]")
    rc = metrics.get("ranking_consistency", {})
    if rc:
        for k, val in rc.items():
            sp = val.get("spearman")
            kt = val.get("kendall_tau")
            print(f"  - {k}: spearman={sp}, kendall_tau={kt}")
    else:
        print("  - (no stats)")

    print("\n[인접 티어 간 KPI 마진]")
    margins = metrics.get("tier_adjacent_kpi_margins", {})
    if margins:
        for k, m in margins.items():
            r1 = m.get("Rising_minus_Rookie")
            r2 = m.get("Super_minus_Rising")
            r3 = m.get("Master_minus_Super")
            print(f"  - {k}: ΔRising-Rookie={r1}, ΔSuper-Rising={r2}, ΔMaster-Super={r3}")
    else:
        print("  - (no margins)")


# -------------------- End-to-End + Evidence --------------------
def run_ssl_ranked_with_evidence(
    raw_csv="combined_dataset.csv",
    out_dir="out",
    # SSL 학습
    epochs=50, batch=512, lr=1e-3, temperature=0.5,
    hidden=512, emb=128, proj_hidden=256, proj_dim=64,
    noise=0.05, mask=0.05, grad_accum=1, save_every=10,
    # Ranked tier cut
    p_master=0.05, p_super=0.15, p_rising=0.30, alpha_ssl=0.5,
    # Exec
    seed=42, cpu=False, workers=4
):
    set_seed(seed)
    device = torch.device("cuda" if torch.cuda.is_available() and not cpu else "cpu")
    print("Device:", device)

    raw_csv = Path(raw_csv)
    out_dir = Path(out_dir); out_dir.mkdir(parents=True, exist_ok=True)
    ssl_dir = out_dir / "ssl"
    tier_dir = out_dir / "ranked_tiers"
    ssl_dir.mkdir(exist_ok=True, parents=True)
    tier_dir.mkdir(exist_ok=True, parents=True)

    # 1) 데이터 로드/정리
    df = pd.read_csv(raw_csv)
    print("Data shape:", df.shape)
    if "price" in df.columns: df["price"] = df["price"].apply(parse_price)
    assert "host_id" in df.columns, "host_id 컬럼이 필요합니다."

    # 2) 리스팅 피처 엔지니어링
    num_cols_cand = [
        "price","number_of_reviews","reviews_per_month","calculated_host_listings_count",
        "minimum_nights","maximum_nights","availability_30","availability_60",
        "availability_90","availability_365","review_scores_rating",
        "review_scores_accuracy","review_scores_cleanliness","review_scores_checkin",
        "review_scores_communication","review_scores_location","review_scores_value"
    ]
    num_cols = [c for c in num_cols_cand if c in df.columns]
    cat_cols_cand = ["room_type","property_type","neighbourhood_cleansed",
                     "bathrooms_text","instant_bookable","host_response_time"]
    cat_cols = [c for c in cat_cols_cand if c in df.columns]

    num_df = df[num_cols].copy().apply(pd.to_numeric, errors="coerce")
    num_df = num_df.fillna(num_df.median())

    oh_parts, meta_top = [], {}
    for c in cat_cols:
        k = 30 if c == "property_type" else (50 if c == "neighbourhood_cleansed" else 10)
        oh, top_vals = topk_one_hot(df[c], k=k, name=c)
        oh_parts.append(oh); meta_top[c] = {"k": k, "top_values": top_vals}
    cat_oh = pd.concat(oh_parts, axis=1) if oh_parts else pd.DataFrame(index=df.index)

    X = pd.concat([num_df, cat_oh], axis=1).astype(float)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X.values)
    pd.DataFrame(Xs, columns=X.columns).to_csv(ssl_dir/"listing_features.csv", index=False)
    dump(scaler, ssl_dir/"scaler.joblib")
    (ssl_dir/"feature_metadata.json").write_text(json.dumps({
        "num_cols": num_cols, "cat_cols": cat_cols, "cat_topk": meta_top,
        "feature_columns": X.columns.tolist()
    }, ensure_ascii=False, indent=2))

    host_ids = df["host_id"].values

    # 3) 대조학습 데이터로더
    ds = ContrastiveListingDataset(Xs, host_ids, feature_noise=noise, mask_prob=mask)
    dl = DataLoader(ds, batch_size=batch, shuffle=True, num_workers=workers,
                    drop_last=True, pin_memory=(device.type=="cuda"))

    # 4) SSL 학습
    input_dim = Xs.shape[1]
    encoder = MLPEncoder(input_dim, hidden, emb).to(device)
    projector = ProjectionHead(emb, proj_hidden, proj_dim).to(device)
    ckpt_dir = ssl_dir / "checkpoints"
    train_contrastive(encoder, projector, dl, device,
                      epochs=epochs, lr=lr, temperature=temperature,
                      grad_accum_steps=grad_accum, save_every=save_every, out_dir=ckpt_dir)

    # 5) 임베딩 → 호스트 평균 임베딩
    encoder.eval(); E=[]
    with torch.no_grad():
        bs = 4096
        for i in range(0, Xs.shape[0], bs):
            xb = torch.from_numpy(Xs[i:i+bs]).to(device).float()
            E.append(encoder(xb).cpu().numpy())
    E = np.vstack(E)                             # (n_listings, emb)
    emb_df = pd.DataFrame(E); emb_df["host_id"] = host_ids
    host_vec = emb_df.groupby("host_id").mean().sort_index()   # (n_hosts, emb)
    host_vec.to_csv(ssl_dir/"host_vectors_ssl.csv")

    # 6) KPI 집계
    host_kpis = build_host_kpis(df)

    # 7) 약한 라벨(슈퍼호스트) 추출(있으면)
    weak_labels = None
    for cand in ["host_is_superhost","is_superhost"]:
        if cand in df.columns:
            weak_labels = df.groupby("host_id")[cand].first()
            break

    # 8) 점수/구성요소 산출
    score, kpi_score, maybe_ssl_score, Z_kpi, Z_contrib, meta = build_score_with_components(
        host_kpis=host_kpis,
        ssl_host_vec=host_vec,
        weak_labels=weak_labels,
        alpha_ssl=alpha_ssl
    )

    # 9) 등급 + 커트라인
    tiers, cutoffs = cut_into_tiers_with_cutoffs(score, p_master=p_master, p_super=p_super, p_rising=p_rising)

    # 10) 결과 저장(최종 테이블: KPI/SSL/최종점수/등급)
    result = pd.DataFrame({
        "host_id": score.index,
        "score": score.values,
        "kpi_score": kpi_score.reindex(score.index).values,
        "tier": tiers.reindex(score.index).values
    }).set_index("host_id")
    if maybe_ssl_score is not None:
        result["ssl_score"] = maybe_ssl_score.reindex(score.index)
    result = result.join(host_kpis, how="left")
    result.reset_index().to_csv(tier_dir/"host_tiers_ranked.csv", index=False)

    # 11) KPI 기여도 저장
    contrib = Z_contrib.copy()
    contrib["kpi_score"] = kpi_score
    contrib["final_score"] = score
    contrib["tier"] = tiers
    contrib.reset_index().to_csv(tier_dir/"kpi_contributions.csv", index=False)

    # 12) 등급별 KPI 평균 + 단조성 체크 + 콘솔 예쁘게 출력
    report, violations = tier_report(host_kpis, tiers)
    report.to_csv(tier_dir/"tier_kpi_report.csv")
    _ = pretty_tier_kpi_table(host_kpis, tiers)

    # 13) 커트라인/분포 메타 저장
    (tier_dir/"score_cutoffs.json").write_text(json.dumps(cutoffs, ensure_ascii=False, indent=2))
    distro = result["tier"].value_counts().to_dict()
    (tier_dir/"tier_distribution.json").write_text(json.dumps(distro, ensure_ascii=False, indent=2))

    # 14) 증거 요약 리포트
    lines = []
    lines.append("# Tiering Evidence Report\n")
    lines.append("## 1) Percentile Cutoffs (최종 점수 기준)\n")
    for k,v in cutoffs.items():
        lines.append(f"- {k}: {v:.6f}")
    lines.append("\n## 2) Tier Distribution (개수)\n")
    for k,v in distro.items():
        lines.append(f"- {k}: {v}")
    lines.append("\n## 3) KPI Weights & Directions\n")
    for k,v in meta["weights"].items():
        sign = "↑(+)" if meta["dir_map"][k] > 0 else "↓(-)"
        lines.append(f"- {k}: weight={v}, dir={sign}")
    lines.append("\n## 4) Monotonicity Check on Tier Averages\n")
    if violations:
        lines.append("⚠ 단조성 위반:")
        for v in violations: lines.append(f"- {v}")
    else:
        lines.append("✅ 주요 KPI 단조성 만족")
    lines.append("\n## 5) Sample Hosts per Tier (Top by Score)\n")
    for t in ["Master","Super","Rising","Rookie"]:
        sub = result[result["tier"]==t].sort_values("score", ascending=False).head(5)
        if len(sub)==0: continue
        lines.append(f"### {t} (상위 5)\n")
        cols = ["host_id","score","kpi_score"]
        if "ssl_score" in sub.columns: cols.append("ssl_score")
        lines.append(sub.reset_index()[cols].to_markdown(index=False))
        lines.append("")
    (tier_dir/"evidence_report.md").write_text("\n".join(lines), encoding="utf-8")

    # 15) 추가: metrics.json 계산/저장 -----------------------------
    metrics = {}

    # 15-1) SSL 표현 품질
    try:
        Hv = host_vec.values.astype(np.float64)
        metrics["ssl_uniformity_log"] = uniformity_metric(Hv, t=2.0, num_samples=min(20000, Hv.shape[0]*10))
    except Exception as e:
        metrics["ssl_uniformity_log"] = None
        metrics["ssl_uniformity_log_error"] = str(e)

    try:
        metrics["ssl_alignment_mean_l2"] = alignment_metric(E.astype(np.float64), host_ids, num_hosts=300)
    except Exception as e:
        metrics["ssl_alignment_mean_l2"] = None
        metrics["ssl_alignment_mean_l2_error"] = str(e)

    # 15-2) 약라벨 기반 평가(존재할 때만)
    if weak_labels is not None and weak_labels.notna().any():
        try:
            metrics.update(topk_precision(score, weak_labels, ks=(0.05, 0.10, 0.20)))
        except Exception as e:
            metrics["topk_precision_error"] = str(e)

        try:
            idx = host_vec.index.intersection(weak_labels.index)
            if len(idx) >= 20:
                X_eval = host_vec.loc[idx].values
                y_eval = weak_labels.loc[idx].apply(bool_from_any).astype(int).values
                ss = StandardScaler().fit_transform(X_eval)
                clf = LogisticRegression(max_iter=1000)
                clf.fit(ss, y_eval)
                prob = clf.predict_proba(ss)[:,1]
                metrics["linear_eval_auroc"] = float(roc_auc_score(y_eval, prob))
                metrics["linear_eval_auprc"] = float(average_precision_score(y_eval, prob))
                metrics["weak_label_rate"] = float(y_eval.mean())
            else:
                metrics["linear_eval_auroc"] = None
                metrics["linear_eval_auprc"] = None
                metrics["weak_label_rate"] = label_rate(weak_labels)
        except Exception as e:
            metrics["linear_eval_error"] = str(e)
    else:
        metrics["weak_label_rate"] = label_rate(weak_labels)

    # 15-3) 티어 구조 품질
    metrics["monotonicity_violations"] = len(violations)
    if len(violations) > 0:
        metrics["monotonicity_violation_list"] = violations

    # 인접 티어 KPI margin
    try:
        order = ["Rookie","Rising","Super","Master"]
        margins = {}
        for c in report.columns:
            vals = report.loc[order, c].values
            if np.any(np.isnan(vals)): continue
            margins[c] = {
                "Rising_minus_Rookie": float(vals[1] - vals[0]),
                "Super_minus_Rising":  float(vals[2] - vals[1]),
                "Master_minus_Super":  float(vals[3] - vals[2]),
            }
        metrics["tier_adjacent_kpi_margins"] = margins
    except Exception as e:
        metrics["tier_adjacent_kpi_margins_error"] = str(e)

    # Silhouette (호스트 임베딩 vs 티어)
    try:
        label_map = {"Rookie":0,"Rising":1,"Super":2,"Master":3}
        y_sil = tiers.map(label_map).reindex(host_vec.index)
        mask = y_sil.notna()
        if mask.sum() >= 10 and len(np.unique(y_sil[mask])) >= 2:
            metrics["silhouette_host_vec_vs_tier"] = float(
                silhouette_score(host_vec.loc[mask].values, y_sil[mask].astype(int).values)
            )
        else:
            metrics["silhouette_host_vec_vs_tier"] = None
    except Exception as e:
        metrics["silhouette_error"] = str(e)

    # 15-4) 랭킹 일관성: Score vs KPI 순위상관
    try:
        rank_corr = {}
        kpi_cols = ["avg_rating","sum_reviews","avg_reviews_per_month","median_price","availability_30","availability_90","availability_365"]
        for c in kpi_cols:
            if c not in host_kpis.columns: continue
            x = score.reindex(host_kpis.index).values
            y = host_kpis[c].values
            mask = (~np.isnan(x)) & (~np.isnan(y))
            if mask.sum() >= 10:
                sp = spearmanr(x[mask], y[mask], nan_policy='omit').correlation
                kt = kendalltau(x[mask], y[mask]).correlation
                rank_corr[c] = {"spearman": float(sp), "kendall_tau": float(kt)}
        metrics["ranking_consistency"] = rank_corr
    except Exception as e:
        metrics["ranking_consistency_error"] = str(e)

    # 저장 + 콘솔 출력
    (tier_dir/"metrics.json").write_text(json.dumps(metrics, ensure_ascii=False, indent=2))
    pretty_print_metrics(metrics)

    # 16) 콘솔 요약
    print("\n=== 완료 ===")
    print("출력 폴더:", out_dir.resolve())
    print(" [SSL]  - listing_features.csv\n   - scaler.joblib\n   - feature_metadata.json\n   - checkpoints/ssl_epoch*.pt\n   - host_vectors_ssl.csv")
    print(" [Tier] - host_tiers_ranked.csv\n   - kpi_contributions.csv\n   - tier_kpi_report.csv\n   - score_cutoffs.json\n   - tier_distribution.json\n   - evidence_report.md\n   - metrics.json")
    print("\n티어 분포:")
    print(pd.Series(distro))
    if violations:
        print("\n⚠ 단조성 위반:")
        for v in violations: print(" -", v)
    else:
        print("\n✅ 주요 KPI 단조성 만족")


# -------------------- CLI --------------------
if __name__ == "__main__":
    ap = argparse.ArgumentParser()
    ap.add_argument("--raw_csv", type=str, default="./combined_dataset.csv")
    ap.add_argument("--out_dir", type=str, default="./out")
    # SSL
    ap.add_argument("--epochs", type=int, default=50)
    ap.add_argument("--batch", type=int, default=512)
    ap.add_argument("--lr", type=float, default=1e-3)
    ap.add_argument("--temperature", type=float, default=0.5)
    ap.add_argument("--hidden", type=int, default=512)
    ap.add_argument("--emb", type=int, default=128)
    ap.add_argument("--proj_hidden", type=int, default=256)
    ap.add_argument("--proj_dim", type=int, default=64)
    ap.add_argument("--noise", type=float, default=0.05)
    ap.add_argument("--mask", type=float, default=0.05)
    ap.add_argument("--grad_accum", type=int, default=1)
    ap.add_argument("--save_every", type=int, default=10)
    # Ranked cut
    ap.add_argument("--p_master", type=float, default=0.05)
    ap.add_argument("--p_super", type=float, default=0.15)
    ap.add_argument("--p_rising", type=float, default=0.30)
    ap.add_argument("--alpha_ssl", type=float, default=0.5)
    # Exec
    ap.add_argument("--seed", type=int, default=42)
    ap.add_argument("--cpu", action="store_true")
    ap.add_argument("--workers", type=int, default=4)

    args, _ = ap.parse_known_args()  # 주피터 인자 무시
    run_ssl_ranked_with_evidence(**vars(args))

```

```python
# -*- coding: utf-8 -*-
# 필요한 표:
#  ① host_id별 등장 횟수(전체)
#  ② 호스트별 보유 숙소 수 분포(=count별 호스트 수)
#  ③ 요약(고유 host 수, 중복 레코드 수, 단일/다중 보유 호스트 수/비율)
#  ④ 상위 N명의 host

import pandas as pd
import numpy as np

# ==============================
# 0) 데이터 로드
# ==============================
PATH = "/content/combined_dataset.csv"   # 필요 시 경로 수정
df = pd.read_csv(PATH)

if "host_id" not in df.columns:
    raise KeyError("데이터에 'host_id' 컬럼이 없습니다.")

# ==============================
# 1) host_id별 등장 횟수 테이블
# ==============================
host_counts = (
    df["host_id"]
    .value_counts(dropna=False)                    # 결측이 있으면 포함
    .rename_axis("host_id")
    .reset_index(name="count")
    .sort_values(["count", "host_id"], ascending=[False, True])
    .reset_index(drop=True)
)

# ==============================
# 2) 요약 통계 (중복 고려)
# ==============================
total_records = len(df)
unique_hosts = df["host_id"].nunique(dropna=False)
duplicate_records = total_records - unique_hosts

num_hosts_single = int((host_counts["count"] == 1).sum())   # 숙소 1개 호스트 수
num_hosts_multi  = int((host_counts["count"] >= 2).sum())   # 숙소 2개 이상 호스트 수

summary = pd.DataFrame({
    "전체 레코드 수": [total_records],
    "고유 host_id 수(=실제 호스트 수)": [unique_hosts],
    "중복된 레코드 수(레코드 관점)": [duplicate_records],
    "1개만 가진 호스트 수": [num_hosts_single],
    "2개 이상 가진 호스트 수": [num_hosts_multi],
    "2개 이상 호스트 비율(%)": [round(100 * num_hosts_multi / unique_hosts, 2) if unique_hosts else np.nan],
})

# ==============================
# 3) 호스트별 보유 숙소 수 분포
#    (count=1,2,3,... 별로 해당 호스트가 몇 명인지)
# ==============================
count_distribution = (
    host_counts["count"]
    .value_counts()            # 특정 count(보유 숙소 수)에 해당하는 host 수
    .sort_index()
    .rename_axis("count_per_host")
    .reset_index(name="num_hosts")
)

# ==============================
# 4) 상위 N명의 host
# ==============================
TOP_N = 10
top_n_hosts = host_counts.head(TOP_N).copy()

# ==============================
# (선택) CSV로 저장
# ==============================
# host_counts.to_csv("host_counts_by_id.csv", index=False)
# count_distribution.to_csv("host_count_distribution.csv", index=False)
# summary.to_csv("host_id_summary.csv", index=False)
# top_n_hosts.to_csv("top_hosts_by_count.csv", index=False)

# ==============================
# 미리보기 (필요 시 출력)
# ==============================
print("\n[요약]")
print(summary.to_string(index=False))

print("\n[host_id별 등장 횟수 상위 10]")
print(top_n_hosts.to_string(index=False))

print("\n[호스트별 보유 숙소 수 분포 (상위 20행 미리보기)]")
print(count_distribution.head(20).to_string(index=False))

# 결과 DataFrame들:
# - host_counts
# - count_distribution
# - summary
# - top_n_hosts
```

```python
# -*- coding: utf-8 -*-
"""
요구사항 반영:
- '모든 칼럼' 비교표 (host_id별 1행 요약)
- 랭킹 기준: host_id 등장 횟수(count) 상위 100
- 결측/문자 처리:
    * 퍼센트(%) 컬럼 → 안전하게 숫자(float) 변환 (90% 이상이 %면 변환)
    * 달러/숫자 문자열( $, , 포함 ) → 안전하게 숫자(float) 변환 (90% 이상이 숫자 형식이면 변환)
    * 슈퍼호스트 t/f → 1/0 매핑 + 원본 보존
- 집계 규칙:
    * 숫자형 기본: 평균(mean)
    * 'number_of_reviews' 등 합계가 의미 있는 일부는 sum (SUM_COLS에서 지정)
    * 문자/범주형: 첫 값(first) 대표값
- 산출물:
    * rank, host_id, count(점수) 포함한 상위 100 비교표 출력
    * CSV 저장: /mnt/data/top100_all_columns_comparison.csv
"""

import re
import numpy as np
import pandas as pd

# =============== 0) 데이터 로드 ===============
PATH = "/content/combined_dataset.csv"
df = pd.read_csv(PATH)

if "host_id" not in df.columns:
    raise KeyError("데이터에 'host_id' 컬럼이 없습니다.")

# =============== 1) 안전한 형 변환 유틸 ===============
def _is_numeric_string_series(s: pd.Series) -> pd.Series:
    """쉼표 제거 후 숫자 형태인지 여부를 각 원소별로 판단"""
    s_str = s.astype(str).str.strip()
    # 숫자 형식 정규식: 부호? 정수 . 소수?
    return s_str.str.match(r"^-?\d+(?:\.\d+)?$")

def _safe_convert_percent(col: pd.Series) -> pd.Series:
    """컬럼의 90% 이상이 %로 끝나면 % 제거 후 float 변환"""
    s = col.dropna().astype(str).str.strip()
    if len(s) == 0:
        return col
    pct_ratio = s.str.endswith("%").mean()
    if pct_ratio > 0.9:
        tmp = col.astype(str).str.strip().str.rstrip("%")
        tmp = tmp.replace({"": np.nan, "nan": np.nan})
        # 쉼표 제거 후 숫자인지 확인
        tmp2 = tmp.str.replace(",", "", regex=False)
        mask_numeric = _is_numeric_string_series(tmp2)
        # 90% 이상 숫자로 해석 가능하면 변환
        if mask_numeric.mean() > 0.9:
            out = pd.to_numeric(tmp2, errors="coerce")
            return out
    return col

def _safe_convert_currency_or_numeric(col: pd.Series) -> pd.Series:
    """
    달러($), 쉼표(,)가 섞인 '숫자 문자열'을 안전하게 float로 변환.
    컬럼의 90% 이상이 숫자 형태로 해석 가능할 때만 변환.
    """
    s = col.dropna().astype(str).str.strip()
    if len(s) == 0:
        return col
    # $, , 제거
    cleaned = s.str.replace("$", "", regex=False).str.replace(",", "", regex=False)
    num_ratio = _is_numeric_string_series(cleaned).mean()
    if num_ratio > 0.9:
        # 전체에 동일 변환 적용
        tmp = col.astype(str).str.strip().str.replace("$", "", regex=False).str.replace(",", "", regex=False)
        tmp = tmp.replace({"": np.nan, "nan": np.nan})
        out = pd.to_numeric(tmp, errors="coerce")
        return out
    return col

# =============== 2) 컬럼별 전처리 (퍼센트/달러/숫자 문자열) ===============
for c in df.columns:
    if df[c].dtype == "object":
        # 1) 퍼센트 컬럼 먼저 시도
        new_col = _safe_convert_percent(df[c])
        # 2) 퍼센트로 안 바뀌었으면, 달러/숫자 문자열 변환 시도
        if new_col.dtype == "object":
            new_col = _safe_convert_currency_or_numeric(new_col)
        df[c] = new_col

# =============== 3) 슈퍼호스트 매핑 (원본 보존) ===============
if "host_is_superhost" in df.columns and "host_is_superhost_num" not in df.columns:
    df["host_is_superhost_num"] = df["host_is_superhost"].map({"t": 1, "f": 0})

# =============== 4) host_id 등장 횟수(점수) 및 Top 100 추출 ===============
host_counts = (
    df["host_id"].value_counts(dropna=False)
      .rename_axis("host_id")
      .reset_index(name="count")
      .sort_values(["count", "host_id"], ascending=[False, True])
      .reset_index(drop=True)
)
topN = 100
top_ids = host_counts.head(topN)["host_id"].tolist()
top_df = df[df["host_id"].isin(top_ids)].copy()

# =============== 5) 모든 칼럼 집계 규칙 구성 ===============
# 합계가 의미 있는 컬럼 목록 (필요에 따라 추가 가능)
SUM_COLS = set([
    "number_of_reviews",
    # 예시) "availability_365", "calculated_host_listings_count" 등 필요시 추가
])

agg_funcs = {}
for col in top_df.columns:
    if col == "host_id":
        continue
    # 합계 우선 규칙
    if col in SUM_COLS:
        agg_funcs[col] = "sum"
        continue
    # 숫자형이면 평균
    if np.issubdtype(top_df[col].dtype, np.number):
        agg_funcs[col] = "mean"
    else:
        # 문자/범주형은 대표값(첫 값)
        agg_funcs[col] = "first"

# =============== 6) 그룹 집계 & count/순위 결합 ===============
agg_top = top_df.groupby("host_id", dropna=False).agg(agg_funcs).reset_index()
# count(점수) 결합
agg_top = agg_top.merge(host_counts, on="host_id", how="left")
# count 내림차순 정렬
agg_top = agg_top.sort_values(["count", "host_id"], ascending=[False, True]).reset_index(drop=True)
# rank 추가
agg_top.insert(0, "rank", range(1, len(agg_top) + 1))

# =============== 7) 표시용 반올림 (숫자형만) ===============
rounded = agg_top.copy()
for c in rounded.select_dtypes(include=[np.number]).columns:
    # rank, count는 정수 유지
    if c in ["rank", "count"]:
        rounded[c] = rounded[c].astype(int)
    else:
        rounded[c] = rounded[c].round(3)

# 상위 100만 유지 (그 이하 데이터가 없을 수 있으니 min으로 방어)
rounded = rounded.head(min(topN, len(rounded)))

# =============== 8) 출력 & 저장 ===============
print("\n=== 상위 Top {} host_id 모든 칼럼 비교표 ===".format(topN))
print(rounded.to_string(index=False))

OUT_CSV = "/content/top100_all_columns_comparison.csv"
rounded.to_csv(OUT_CSV, index=False)
print("\n▶ 저장 완료:", OUT_CSV)

# 결과 DataFrame:
# - rounded (표시용 반올림 포함)
# - agg_top (원본 스케일 요약)

```
